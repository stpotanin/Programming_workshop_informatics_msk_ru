```r
# @title https://stepik.org/lesson/244349/step/10
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
# lines <- c('1', '0', '7', '0', '9', '0', '0') # 17

a <- sapply(lines, as.integer)
for (i in 1:(length(a)-1)) {
    if (!a[i] & !a[i+1]) {
        threshold <- i-1
        break
    }
}

cat(sum(a[1:threshold]))
```

```r
# @title https://stepik.org/lesson/244349/step/11
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
# lines <- c('1', '7', '7', '9', '1', '0') # 2

a <- sapply(lines, as.integer)
n <- match(0, a)-1
m <- rep(1, n)

if (n > 1) {
    for (i in 2:n) {
        if (a[i] == a[i-1]) {
            m[i] <- m[i-1] + 1
        }
    }
}

cat(max(m))
```

```r
# @title https://stepik.org/lesson/244349/step/12
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
# lines <- c('3', '3', '3', '3', '3', '0') # 1
# lines <- c('3', '1', '3', '3', '3', '0') # 2
# lines <- c('3', '1', '2', '3', '3', '0') # 3

a <- sapply(lines, as.integer)
n <- match(0, a)-1
m <- rep(1, n)

if (n > 1) {
    for (i in 2:n) {
        if (i > 2 & (a[i-1] < a[i] & a[i-2] < a[i-1] || a[i-1] > a[i] & a[i-2] > a[i-1])) {
            m[i] <- m[i-1] + 1
        } else if ((a[i-1] < a[i]) || (a[i-1] > a[i])) {
            m[i] <- 2
        }
    }
}

cat(max(m))
```

```r
# @title https://stepik.org/lesson/244349/step/13
# https://informatics.msk.ru/mod/statements/view.php?chapterid=3079
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
# lines <- c('1', '2', '1', '2', '1', '0') # 2

a <- sapply(lines, as.integer)
n <- match(0, a)-1  # длина последовательности
m <- 0

for (i in 2:(n-1)) {
    if (a[i-1] < a[i] & a[i] > a[i+1])
        m <- m + 1
}

cat(m)
```

```r
# @title https://stepik.org/lesson/244349/step/14
# https://informatics.msk.ru/mod/statements/view.php?chapterid=3080
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
# lines <- c('1', '2', '1', '1', '2', '1', '2', '1', '0') # 2

a <- sapply(lines, as.integer)
n <- match(0, a)-1  # длина последовательности
m <- c()
# находим и сохраняем индексы максимумов
for (i in 2:(n-1)) {
    if (a[i-1] < a[i] & a[i] > a[i+1])
        m <- c(m, i)
}
# считаем расстояния между максимумами и выводим минимальное
if (length(m) < 2) cat(0) else cat(min(m[2:length(m)] - m[1:(length(m)- 1)]))
```
