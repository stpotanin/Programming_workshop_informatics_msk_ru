```python
def gen_gnche(n: int)-> int:
    '''return 1 2 2 3 3 3 4 4 4 4 5...'''
    i, m = 1, 0
    while i <= n:
        m, k = m + 1, 0
        while i <= n and k < m:
            yield m
            i, k = i + 1, k + 1

print(*gen_gnche(int(input())))
```

```python
mode = None
prev = int(input())
max_cnt = cnt = 1

for n in map(int, iter(input, '0')):
    if n == prev:
        cnt = 1
        mode = None
    elif n > prev:
        cnt = cnt + 1 if mode == 'ASC' else 2
        mode = 'ASC'
    else:
        cnt = cnt + 1 if mode == 'DESC' else 2
        mode = 'DESC'
    max_cnt = max(cnt, max_cnt)
    prev = n

print(max_cnt)
```

```python
# Функция для НОД
gcd = lambda a, b: abs(a) if b == 0 else gcd(b, a % b)

# Функция для решения с возможностью замены имён сосудов
def solver(A_cap, B_cap, target, A_name='A', B_name='B'):
    A, B = 0, 0
    steps = []

    # Наполняем сосуд A (первый)
    A = A_cap
    steps.append(f'>{A_name}')

    while A != target and B != target:
        # Переливаем из А в B либо то, что есть, либо то, что влезет
        transfer = min(A, B_cap - B)
        A -= transfer
        B += transfer
        steps.append(f'{A_name}>{B_name}')

        if A == target or B == target:
            break

        # Если сосуд B (второй) наполнился, то опустошаем
        if B == B_cap:
            B = 0
            steps.append(f'{B_name}>')

        # Если сосуд A (первый) опустошился, то наполняем
        if A == 0:
            A = A_cap
            steps.append(f'>{A_name}')

    return steps

# Ввод данных
a, b, n = int(input()), int(input()), int(input())

# Если есть решение, то создаём два для выбора лучшего
if n % gcd(a, b) != 0 or n > max(a, b):
    print("Impossible")
else:
    steps1 = solver(a, b, n, A_name='A', B_name='B')
    steps2 = solver(b, a, n, A_name='B', B_name='A')

    # Выбираем вариант с меньшим числом шагов
    if len(steps1) >= len(steps2):
        for s in steps1:
            print(s)
    else:
        for s in steps2:
            print(s)
```
