```r
# @title https://stepik.org/lesson/463039/step/2
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
lines <- c('Abrakadabra') # вывод:
# r
# r
# Abrak
# Abrakadab
# Arkdba
# baaar
# arbadakarbA
# abdkrA
# 11

s <- unlist(strsplit(lines[1], ''))

cat(s[3], sep='', end='\n')         # третий символ этой строки
cat(rev(s)[2], sep='', end='\n')    # предпоследний символ этой строки
cat(s[1:5], sep='', end='\n')       # первые пять символов этой строки
cat(s[1:(length(s)-2)], sep='', end='\n')         # вся строка, кроме последних двух
cat(s[seq(1, length(s), by=2)], sep='', end='\n') # все символы с четными индексами (индексация с 0)
cat(s[seq(2, length(s), by=2)], sep='', end='\n') # с нечетными индексами (индексация с 0)
cat(rev(s), sep='', end='\n')                     # все в обратном порядке
cat(rev(s)[seq(1, length(s), by=2)], sep='', end='\n') # все через один в обратном порядке
cat(length(s))                      # длина данной строки
```

```r
# @title https://stepik.org/lesson/463039/step/6
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
# lines <- c('comfort') # -1
# lines <- c('coffee') # 3
# lines <- c('cat') # -2

s <- unlist(strsplit(lines[1], ''))

# проверяем первое вхождение
if (!'f' %in% s) cat(-2) else {
    # удаляем первое вхождение
    index <- match('f', s)
    s <- s[-index]
    # проверяем второе вхождение
    if (!'f' %in% s) cat(-1) else cat(match('f', s))
}
```

```r
# @title https://stepik.org/lesson/463039/step/7
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
# lines <- c('In the hole in the ground there lived a hobbit') # In tobbit

s <- unlist(strsplit(lines[1], ''))
res1 <- s[1:match('h', tolower(s))-1]
s <- rev(s)
res2 <- rev(s[1:match('h', tolower(s))-1])

cat(res1, res2, sep='')
```

```r
# @title https://stepik.org/lesson/463039/step/8
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
# lines <- c('In the hole in the ground there lived a hobbit') # In th a devil ereht dnuorg eht ni eloh ehobbit

s <- unlist(strsplit(lines[1], ''))

left <- min(unlist(gregexpr('h', tolower(lines[1]))))
right <- max(unlist(gregexpr('h', tolower(lines[1]))))

cat(s[1:left], rev(s[(left+1):(right-1)]), s[right:length(s)], sep='')
```

```r
# @title https://stepik.org/lesson/463039/step/12
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
# lines <- c('In the hole in the ground there lived a hobbit') # In the Hole in tHe ground tHere lived a hobbit

s <- unlist(strsplit(lines[1], ''))

left <- min(unlist(gregexpr('h', tolower(lines[1]))))
right <- max(unlist(gregexpr('h', tolower(lines[1]))))

cat(s[1:left], gsub('h', 'H', s[(left+1):(right-1)]), s[right:length(s)], sep='')
```
