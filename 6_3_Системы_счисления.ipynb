```r
# @title https://stepik.org/lesson/464318/step/2 Из двоичной в шестнадцатиричную
bin_to_hex <- function(x) {
    if (x == '0000') return('0')
    else if (x == '0001') return('1')
    else if (x == '0010') return('2')
    else if (x == '0011') return('3')
    else if (x == '0100') return('4')
    else if (x == '0101') return('5')
    else if (x == '0110') return('6')
    else if (x == '0111') return('7')
    else if (x == '1000') return('8')
    else if (x == '1001') return('9')
    else if (x == '1010') return('A')
    else if (x == '1011') return('B')
    else if (x == '1100') return('C')
    else if (x == '1101') return('D')
    else if (x == '1110') return('E')
    else if (x == '1111') return('F')
}

lines <- readLines(file('stdin'))

# проверка
# lines <- c('101') # 5
# lines <- c('1101010101010') # 1AAA

a <- unlist(strsplit(lines[1], ''))

# добавляем нули слева до кратности четырём
a_0 <- c(rep('0', ((4 - length(a) %% 4) %% 4)), a)

# матрица по 4 символа
m <- matrix(a_0, ncol = 4, byrow = TRUE)

# лепим ответ
res <- ''
for (i in 1:nrow(m)) {
    symbol <- paste(m[i,], sep = '', collapse = '')
    res <- paste(res, bin_to_hex(symbol), sep = '')
}

cat(res)
```

```r
# @title https://stepik.org/lesson/464318/step/3 Universal convertor
lines <- readLines(file('stdin'))

# проверка
lines <- c('10', '19', '2') # 10011
# lines <- c('10', '32', '3') # 1012

n <- as.integer(lines[1])
a_list <- unlist(strsplit(lines[2], '')) # список из цифр
digits <- length(a_list)                 # число цифр в числе
k <- as.integer(lines[3])

# цифры с буквами
numbers <- c(as.character(0:9), LETTERS)

x <- 0 # десятичный эквивалент
for (i in 1:digits) {
    weight <- which(numbers == a_list[i]) - 1 # вес символа
    x <- x + weight * n^(digits - i)
}

res_list <- c() # список цифр результата
while (x > 0) {
    weight <- x %% k # остаток
    res_list <- c(res_list, numbers[weight + 1])
    x <- x %/% k # целое частное
}

cat(rev(res_list), sep = '')
```

```r
# @title https://stepik.org/lesson/464318/step/4 Two's complement - 1 (ЕСЛИ А - n-ричное)
lines <- readLines(file('stdin'))

# проверка
# lines <- c('3', '8') # 00000011
lines <- c('57', '8') # 00111001
lines <- c('-57', '8') # 111010001 ?

a_list <- unlist(strsplit(lines[1], ''))
n <- as.integer(lines[2])

# цифры с буквами
numbers <- c(as.character(0:9), LETTERS)

# сохраняем знак и заменяем на ноль
negative <- FALSE
if (a_list[1] == '-') {
    negative <- TRUE
    a_list[1] <- '0'
} else {
    a_list <- c('0', a_list)
}

digits <- length(a_list)# число цифр в числе


# если отрицательное
if (negative) {
    a_list_inv <- c()
    for (i in 1:digits) {
        digit_value <- which(numbers == a_list[i]) - 1
        inv_value <- (n - 1) - digit_value  # дополнение до (n-1)
        a_list_inv <- c(a_list_inv, numbers[inv_value + 1])  # индексация в R с 1
    }
    # cat('a_list_inv =', a_list_inv, '\n')
    a_list <- a_list_inv
}



x <- 0 # десятичный эквивалент
for (i in 1:digits) {
    weight <- which(numbers == a_list[i]) - 1 # вес символа
    x <- x + weight * n^(digits-i)
}


# дополнительный код
if (negative) x <- 2^n - x + 1

# cat('x =', x, '\n')

res_list <- c()  # двоичный эквивалент
while (x > 0) {
    weight <- x %% 2 # остаток
    res_list <- c(res_list, weight)
    x <- x %/% 2 # целое частное
}

res_list <- rev(res_list)

cat(res_list, sep = '')

```

```r
# @title https://stepik.org/lesson/464318/step/6 Инопланетяне
# https://informatics.msk.ru/mod/statements/view.php?id=57571

# функция для перевода в десятичную
dec <- function(a, base) {
    a_list <- unlist(strsplit(as.character(a), ''))
    x <- 0
    for (d in a_list) x <- x * base + which(numbers == d) - 1
    x
}

# алфавит цифры с буквами
numbers <- c(as.character(0:9), LETTERS)

# читаем входную строку
lines <- readLines(file('stdin'))

# проверка
# lines <- c('2+2=4') # 5
# lines <- c('12+3=15') # 6
# lines <- c('F+1=10') # 16
# lines <- c('1+2=10') # 3
# lines <- c('2+3=10') # 5
# lines <- c('0+0=0') # 5

# получаем операнды из строки
a <- unlist(strsplit(lines[1], '[+=]'))

# находим макс. цифру в выражении
a_list <- unlist(strsplit(a, ''))
max_digit <- max(sapply(a_list, function(x) which(numbers == x) - 1))

# мин. основание системы счисления
min_base <- max(max_digit + 1, 2)

# поиск основания для правильности выражения
while (min_base <= 36) {
    if (dec(a[1], min_base) + dec(a[2], min_base) == dec(a[3], min_base)) break
    min_base <- min_base + 1
}

# вывод с проверкой на правильность выражения
cat(ifelse(min_base < 36,
           min_base,
           -1))
```

```r
# @title https://stepik.org/lesson/464318/step/7 Домашнее задание

# алфавит (цифры с буквами)
numbers <- c(as.character(0:9), LETTERS)

# функция для перевода из base-ричной в десятичную
base_to_dec <- function(x, base) {
    x_list <- unlist(strsplit(as.character(x), ''))
    y <- 0
    for (d in x_list) y <- y * base + which(numbers == d) - 1
    y
}

# функция для перевода из десятичной в base-ричную
dec_to_base <- function(x, base = 2) {
    y <- c()
    while (x > 0) {
        y <- c(x %% base, y) # присоединяем слева остаток от деления
        x <- x %/% base # целое частное
    }
    paste(y, collapse = '')
}

dec_to_bin <- function(x) {
    if (x == 0) return("0")
    res <- character(0)
    while (x > 0) {
        res <- c(as.character(x %% 2), res)
        x <- x %/% 2
    }
    paste(res, collapse = "")
}

lines <- readLines(file('stdin'))

# проверка
# lines <- c('101 2 10') # 5
# lines <- c('200 2 10') # -1
# lines <- c('FF 16 2') # 11111111

# получаем операнды из строки
a <- unlist(strsplit(lines[1], ' '))
s1 <- as.integer(a[2])
s2 <- as.integer(a[3])
x <- base_to_dec(a[1], s1) # сразу в десятичную

# находим макс. цифру в числе
x_list <- unlist(strsplit(a[1], ''))
max_digit <- max(sapply(x_list, function(x) which(numbers == x) - 1))

# ответ
cat(ifelse(max_digit < s1, dec_to_base(x, s2), -1))
```

```r
# @title https://stepik.org/lesson/464318/step/8 Binary fraction to decimal
# https://informatics.msk.ru/mod/statements/view3.php?chapterid=728

# функция для перевода из n-ричной в десятичную
base_to_dec <- function(x, base, frac=FALSE) {
  digits <- as.numeric(strsplit(x, "")[[1]])
  exps <- if (frac) {
    -(1:length(digits))
  } else {
    (length(digits) - 1):0
  }
  sum(digits * base ^ exps)
}

lines <- readLines(file('stdin'))

# проверка
lines <- c('0.11') # 0.75
# lines <- c('0.111') # 0.875
# lines <- c('1111111111.1111111111') # 1023.9990234375

digits <- unlist(strsplit(lines[1], '\\.'))

int_part <- ifelse(length(digits) >= 1, digits[1], "0")
frac_part <- ifelse(length(digits) == 2, digits[2], "")
n_frac <- length(unlist(strsplit(frac_part, '')))

res <- base_to_dec(int_part, 2, frac = FALSE) + base_to_dec(frac_part, 2, frac = TRUE)
options(digits = 20)
cat(format(ifelse(res == floor(res), floor(res), res), scientific = FALSE))
```

```r
# @title https://stepik.org/lesson/464318/step/9 Binary periodical fraction to rational number TODO
# https://informatics.msk.ru/mod/statements/view.php?chapterid=732
# функция для перевода из двоичной в десятичную
bin_to_dec <- function(x) sum(as.integer(rev(strsplit(as.character(x), '')[[1]])) *
    2^(seq_along(strsplit(as.character(x), '')[[1]]) - 1))

# функция для НОД
gcd <- function(a, b) ifelse(b == 0, abs(a), gcd(b, a %% b))

lines <- readLines(file('stdin'))

# проверка
# lines <- c('0.01') # 5 2
# lines <- c('10.1') # 5 2
# lines <- c('10.1(01)') # 8 3
# lines <- c('10.(1)') # 3 1
# lines <- c('10') # 2 1

# получаем операнды из строки
digits <- c(unlist(strsplit(lines[1], '\\.')), '0') # '0' на случай целого
whole <- digits[1]
rest <- digits[2]
# нужен список из символов rest
rest_list <- strsplit(digits[2], '')[[1]]

if ('(' %in% rest_list) {
    open <- which(rest_list == '(')
    close <- which(rest_list == ')')
    # проверяем наличие препериода
    is_preperiod <- open > 1
    if (is_preperiod) {
        preperiod <- rest_list[1:(open - 1)]
    } else {
        preperiod <- c()
    }
    period <- rest_list[(open + 1):(close - 1)]
} else {
    preperiod <- rest
    period <- c()
}

if (length(period) > 0) {
    # числитель дробной часть это: concat(preperiod + period) - preperiod
    numerator <- bin_to_dec(paste0(c(preperiod, period), collapse = '')) -
                 ifelse(is_preperiod, bin_to_dec(preperiod), 0)
    denominator <- bin_to_dec(paste0(c(rep('1', length(period)), rep('0', length(preperiod))), collapse = ''))
    # добавляем целую часть
    numerator <- numerator + bin_to_dec(whole) * denominator
} else {
    numerator <- bin_to_dec(preperiod)
    denominator <- 2 ** length(rest_list)
    numerator <- numerator + bin_to_dec(whole) * denominator
}

g <- gcd(numerator, denominator)
numerator <- numerator %/% g
denominator <- denominator %/% g

cat(numerator, denominator)
```
