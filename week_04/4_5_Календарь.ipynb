```r
# @title https://stepik.org/lesson/427802/step/1 вариант
lines <- readLines(file("stdin"))

# для проверки
# lines <- c('90')

n <- as.integer(lines[1])

year <- c(31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31)
month <- 1
day <- n
while (sum(year[1:month]) < n) {
    day <- day - year[month]
    month <- month + 1
}

cat(day, month)
```

```r
# @title https://stepik.org/lesson/427802/step/2
lines <- readLines(file("stdin"))

# для проверки
# lines <- c('1 8 2009') # 3 8 2009
# lines <- c('31 12 2028') # 2 1 2029

date <- as.integer(unlist(strsplit(lines[1], ' ')))

year <- c(31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31)
if ((!(date[3] %% 4) & (date[3] %% 100)) | !(date[3] %% 400)) year[2] <- 29
next_day <- date[1] + sum(year[1:date[2]]) - year[date[2]] + 2
next_day <- next_day %% sum(year)

calendar = function(n) {
    month <- 1
    day <- n
    while (sum(year[1:month]) < n) {
        day <- day - year[month]
        month <- month + 1
    }
    return (c(day, month))
}

d <- calendar(next_day)[1]
m <- calendar(next_day)[2]
y <- ifelse((m < date[2]), date[3]+1, date[3])

cat(d, m, y)
```

```r
# @title https://stepik.org/lesson/427802/step/3
lines <- readLines(file("stdin"))

# для проверки
# lines <- c('02010001') # 2
# lines <- c('29020004') # 1155

d <- as.integer(unlist(strsplit(lines[1], '')))

day <- d[1] * 10 + d[2]
month <- d[3] * 10 + d[4]
year <- d[5] * 1000 + d[6] * 100 + d[7] * 10 + d[8]

dm <- c(31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31)

leap <- function(y) (y %% 4 == 0 & y %% 100 != 0) | (y %% 400 == 0)

years_days <- (year - 1) * 365
leap_days <- if (year > 1) sum(sapply(1:(year - 1), leap)) else 0
indices <- if (month > 1) 1:(month - 1) else 0
month_days <- sum(dm[indices])

if (month > 2 && leap(year)) {
  month_days <- month_days + 1
}

s <- day + month_days + years_days + leap_days
cat(s)
```
