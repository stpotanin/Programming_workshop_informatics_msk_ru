```r
# @title https://stepik.org/lesson/244182/step/2
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
# lines <- c('1', '5', '1') # 10
# lines <- c('3', '1', '4') # 3
# lines <- c('3', '1','0') # 0
# lines <- c('3', '1', '2') # 2

k <- as.integer(lines[1])
m <- as.integer(lines[2])
n <- as.integer(lines[3])

if (n %% k == 0) {
    cat(n %/% k * m * 2)
} else if (n < k) {
    cat(m * 2)
} else {
    cat(n %/% k * m * 2 + m)
}
```

```r
# @title https://stepik.org/lesson/244182/step/6
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
# lines <- c('1', '0', '0') # 0
# lines <- c('-1', '10', '11') # -1 11
# lines <- c('2', '-4', '-6') # -1 3
# lines <- c('2', '1', '3') # ничего не выводится

a <- as.integer(lines[1])
b <- as.integer(lines[2])
c <- as.integer(lines[3])
res <- c()

D <- b^2 - 4 * a * c

if (D == 0) {
    x <- -b / (2 * a)
    if (x == floor(x)) {  # проверка на целочисленность
        res <- c(res, x)
    }
} else if (D > 0) {  # D > 0
    sqrtD <- sqrt(D)
    x1 <- (-b + sqrtD) / (2 * a)
    x2 <- (-b - sqrtD) / (2 * a)
    if (x1 == floor(x1)) {
        res <- c(res, x1)
    }
    if (x2 == floor(x2)) {
        res <- c(res, x2)
    }
}

if (length(res) > 0) {
    cat(sort(res))
}
```

```r
# @title https://stepik.org/lesson/244182/step/7
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
# lines <- c('1', '0', '0') # impossible
# lines <- c('1', '2', '3') # impossible
# lines <- c('3', '4', '5') # right
# lines <- c('3', '4', '4') # acute
# lines <- c('3', '4', '6') # obtuse

t <- sort(c(as.integer(lines[1]), as.integer(lines[2]), as.integer(lines[3])))
if (t[1] <= 0 || t[2] <= 0 || t[3] <= 0 || t[1] + t[2] <= t[3]) {
    cat('impossible')
} else if (t[1]^2 + t[2]^2 == t[3]^2) {
    cat('right')
} else if (t[1]^2 + t[2]^2 > t[3]^2) {
    cat('acute')
} else {
    cat('obtuse')
}
```

```r
# @title https://stepik.org/lesson/244182/step/9
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
# lines <- c('129') # 0 1 2
# lines <- c('119') # 0 0 2
# lines <- c('28') # 8 2 0

n <- as.integer(lines[1])

# сначала вариант без лишних билетов
x60 <- n %/% 60
x10 <- (n %% 60) %/% 10
x1 <- (n %% 10) %/% 1

# проверяем, не выгоднее ли купить лишние поездки
if (x1 * 15 + x10 * 125 + x60 * 440 >= (x60 + 1) * 440) {
    x60 <- x60 + 1
    x10 <- 0
    x1 <- 0
} else if (x1 * 15 + x10 * 125 >= (x10 + 1) * 125) {
    x10 <- x10 + 1
    x1 <- 0
}

cat(x1, x10, x60)
```

```r
# @title https://stepik.org/lesson/244182/step/10
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
# lines <- c('1') # 1 0 0 0 0
# lines <- c('16') # 1 1 1 0 0

n <- as.integer(lines[1])

p <- c(15, 70, 125, 230, 440)
k <- c(1, 5, 10, 20, 60)

variants <- rbind(
  c(0, 0, 0, 0, (n %/% k[5] + 1)),
  c(0, 0, 0, (n %% k[5] %/%k[4] + 1), n %/% k[5]),
  c(0, 0, n %% k[5] %%k [4] %/% k[3] + 1, (n %% k[5] %/%k[4]), n %/% k[5]),
  c(0, n %% k[5] %% k[4] %% k[3] %/% k[2] + 1, n %% k[5] %%k [4] %/% k[3], (n %% k[5] %/%k[4]), n %/% k[5]),
  c(n %% k[5] %%k [4] %% k[3] %% k[2] %/% k[1], n %% k[5] %% k[4] %% k[3] %/% k[2], n %% k[5] %%k [4] %/% k[3], (n %% k[5] %/% k[4]), n %/% k[5])
)

# multiplied_variants <- sweep(variants, MARGIN = 2, STATS = p, FUN = `*`) или
multiplied_variants <- variants * matrix(p, nrow = nrow(variants), ncol = length(p), byrow = TRUE)

cat(variants[which.min(rowSums(multiplied_variants)), ])
```

Кроме того, Цифры V, L и D вообще не повторяются — ни подряд, ни с пропусками. В римской системе они используются максимум один раз в числе. То есть, к примеру, сочетание MCM - допустимо, а VIV - нет.

```r
# @title https://stepik.org/lesson/244182/step/14
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
# lines <- c('4') # IV
lines <- c('9') # IX
# lines <- c('2600') # MMDC
# lines <- c('1987') # MCMLXXXVII
# lines <- c('2449') # MMCDXLIX
# lines <- c('3999') # MMMCMXCIX

x <- as.integer(lines[1])

M <- x %/% 1000
r <- x %% 1000  # остаток
D <- r %/% 500
r <- r %% 500
C <- r %/% 100
r <- r %% 100
L <- r %/% 50
r <- r %% 50
X <- r %/% 10
r <- r %% 10
V <- r %/% 5
r <- r %% 5
I <- r

# res <- paste(M, D, sep="")
# cat(res)

res <- ''

res <- paste(res, strrep('M', M), strrep('D', D), sep = '')

if (C < 4) res <- paste(res, strrep('C', C), sep = '') else res <- paste(res, 'CD', sep = '')

res <- paste(res, strrep('L', L), sep = '')

if (X < 4) res <- paste(res, strrep('X', X), sep = '') else res <- paste(res, 'XL', sep = '')

res <- paste(res, strrep('V', V), sep = '')

if (I < 4) res <- paste(res, strrep('I', I), sep = '') else res <- paste(res, 'IV', sep = '')

cat(res)
```
