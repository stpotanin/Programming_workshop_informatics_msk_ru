```r
# @title https://stepik.org/lesson/427797/step/2
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
# lines <- c('7', 'ADD 5', 'ADD 7', 'COUNT', 'PRESENT 3', 'PRESENT 5', 'ADD 3', 'COUNT') # 2 NO YES 3 в столбик

s <- c()
for (string in lines[-1]) {
  x <- strsplit(string, ' ')[[1]]
    if (x[1] == 'COUNT') cat(length(s), sep='\n')
    if (x[1] == 'ADD') s <- unique(c(s, c(x[2])))
    if (x[1] == 'PRESENT') {
        if (x[2] %in% s) cat('YES', sep='\n') else cat('NO', sep='\n')
    }
}
```

```r
# @title https://stepik.org/lesson/427797/step/3
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
# lines <- c('1 3 2', '4 3 2') # 2
# lines <- c('1 1 1', '2 2 2') # 0

cat(length(which(duplicated(unlist(lapply(strsplit(lines, ' '), unique))))))
```

```r
# @title https://stepik.org/lesson/427797/step/4
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
# lines <- c('1 3 2', '4 3 2') # 2 3
# lines <- c('1 22 3 44', '44 3 22 1') # 1 3 22 44

s <- as.integer(unlist(lapply(strsplit(lines, ' '), unique)))
cat(sort(s[which(duplicated(s))]))
```

```r
# @title https://stepik.org/lesson/427797/step/5
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
# lines <- c('1 2 3 2 3 4') # NO NO NO YES YES NO в столбик

s = unlist(strsplit(lines[1], ' '))
res <- c('NO')
for (i in 2:length(s)) {
    if (s[i] %in% s[1:i-1]) res <- c(res, 'YES') else res <- c(res, 'NO')
}
cat(res, sep='\n')
```
