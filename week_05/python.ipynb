````python
class Heap:
    def __init__(self, arr=None):
        self.arr = arr if arr is not None else []
        self.size = len(self.arr)

    def heapify(self, n, i):
        largest = i
        left = 2 * i + 1
        right = 2 * i + 2

        if left < n and self.arr[left] > self.arr[largest]:
            largest = left
        if right < n and self.arr[right] > self.arr[largest]:
            largest = right

        if largest != i:
            self.arr[i], self.arr[largest] = self.arr[largest], self.arr[i]
            self.heapify(n, largest)

    # Max-куча из массива
    def build_heap(self):
        n = self.size
        for i in range(n // 2 - 1, -1, -1):
            self.heapify(n, i)

    # Сортировка массива in-place сортировкой кучей
    def heapsort(self):
        n = self.size
        self.build_heap()
        for i in range(n - 1, 0, -1):
            # Перемещаем текущий корень (максимум) в конец
            self.arr[0], self.arr[i] = self.arr[i], self.arr[0]
            # Вызываем heapify для уменьшенной кучи
            self.heapify(i, 0)
        return self.arr

_, a = input(), [int(x) for x in input().split()]

h = Heap(a)
sorted_a = h.heapsort()
print(*sorted_a)

# 5 3 4 2 1
````

Определения. Каждый пункт включает в себя свойства последующих, но не наоборот.

1. Бинарное дерево. У каждой вершины не более двух потомков. При этом может быть неполным, т.е. число вершин с одним потомком может быть больше 1.
2. Полное бинарное дерево. Полнота бинарного дерева означает, что все уровни, кроме, возможно, последнего, полностью заполнены, а последний уровень заполнен слева направо без пропусков.
3. Куча (макс или мин), по дефолту макс-куча (max-heap). Полное бинарное дерево с heap-свойством: у каждого узла значение не меньше значений потомков, поэтому максимум в корне.

````python
# @title https://stepik.org/lesson/263808/step/2 Хипуй!
# https://foxford.ru/wiki/informatika/kucha-heap

# Добавление элемента в кучу
def add(Heap, elem):
    Heap.append(elem)
    i = len(Heap) - 1
    while i > 1 and Heap[i // 2] < elem:
        Heap[i] = Heap[i // 2]
        i //= 2
        Heap[i] = elem

# Удаление элемента из кучи
def pop(Heap):
    if len(Heap) == 2:
        return Heap.pop()
    retval = Heap[1]
    Heap[1] = Heap.pop()
    i = 1
    while 2 * i + 1 < len(Heap) \
          and Heap[i] < max(Heap[2 * i], Heap[2 * i + 1]):
        if Heap[2 * i] > Heap[2 * i + 1]:
            Heap[i], Heap[2 * i] = Heap[2 * i], Heap[i]
            i = 2 * i
        else:
            Heap[i], Heap[2 * i + 1] = Heap[2 * i + 1], Heap[i]
            i = 2 * i + 1
    if 2 * i == len(Heap) - 1 and Heap[i] < Heap[2 * i]:
        Heap[i], Heap[2 * i] = Heap[2 * i], Heap[i]
    return retval

Heap = [None,]
n = int(input())
commands = [list(map(int, input().split())) for _ in range(n)]
for command in commands:
    if command[0] == 0:
        add(Heap, command[1])
    else:
        print(pop(Heap))


# In:
# 2
# 0 10000
# 1

# Out:
# 1000
````

````python
# @title https://stepik.org/lesson/263808/step/3 Минимум на отрезке
# Задача Найти минимум в каждом окне длины K в массиве a[0..N-1].
from collections import deque

def sliding_minimum(a, k):
    dq = deque()    # здесь храним индексы
    result = []     # здесь храним минимумы

    for i, x in enumerate(a):
        # Удаляем из хвоста все элементы, больше или равные текущему
        while dq and a[dq[-1]] >= x:
            dq.pop()

        # Добавляем индекс текущего элемента в деку
        dq.append(i)

        # Удаляем из головы то, что вышло за окно
        if dq[0] <= i - k:
            dq.popleft()

        # Когда окно "дозрело", пишем минимум
        if i >= k - 1:
            result.append(a[dq[0]])

    return result


n, k = map(int, input().split())
arr = list(map(int, input().split()))

print(*sliding_minimum(arr, k), sep='\n')



# TODO
# In:
# 7 3
# 1 3 2 4 5 3 1

# Out:
# 1 2 2 3 1 в столбик
````

````python
# @title https://stepik.org/lesson/263810/step/2 Объединение последовательностей
# https://ru.wikipedia.org/wiki/%D0%A1%D0%BE%D1%80%D1%82%D0%B8%D1%80%D0%BE%D0%B2%D0%BA%D0%B0_%D1%81%D0%BB%D0%B8%D1%8F%D0%BD%D0%B8%D0%B5%D0%BC

def merge_sort(arr):
    if len(arr) <= 1:
        return arr

    mid = len(arr) // 2
    left_half = arr[:mid]
    right_half = arr[mid:]

    left_half = merge_sort(left_half)
    right_half = merge_sort(right_half)

    return merge(left_half, right_half)

def merge(left, right):
    merged = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            merged.append(left[i])
            i += 1
        else:
            merged.append(right[j])
            j += 1
    merged.extend(left[i:])
    merged.extend(right[j:])
    return merged

arr = [5, 2, 4, 1, 3]
sorted_arr = merge_sort(arr)
print(*sorted_arr)
````
