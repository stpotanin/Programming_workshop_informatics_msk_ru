```r
# @title https://stepik.org/lesson/244032/step/3
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
# lines <- c('110') # 1
lines <- c('-232') # -232

x <- abs(as.integer(lines[1]))
a <- sprintf("%04d", x)
a_rev <- paste0(substr(rep(a, n <- nchar(a)), n:1, n:1), collapse="")

if (a == a_rev) {
    cat(1)
} else {
    cat(a_rev)
}
```

```r
# @title https://stepik.org/lesson/244032/step/10
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
lines <- c('4', '4', '5', '5') # YES
# lines <- c('1', '1', '3', '3') # NO

a <- as.integer(lines[1])
b <- as.integer(lines[2])
c <- as.integer(lines[3])
d <- as.integer(lines[4])

if (abs(a - c) %in% c(0, 1) && abs(b - d) %in% c(0, 1) && !(a == c && b == d)) {
    cat('YES')
} else {
    cat('NO')
}
```

```r
# @title https://stepik.org/lesson/244032/step/11
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
# lines <- c('1', '1', '3', '2') # YES
# lines <- c('1', '1', '3', '3') # NO

a <- as.integer(lines[1])
b <- as.integer(lines[2])
c <- as.integer(lines[3])
d <- as.integer(lines[4])

# Возможные поля для для хода конём
x1 = c(a+1, b+2)
x2 = c(a+1, b-2)
x3 = c(a-1, b+2)
x4 = c(a-1, b-2)
x5 = c(a+2, b+1)
x6 = c(a+2, b-1)
x7 = c(a-2, b+1)
x8 = c(a-2, b-1)

# Объединяем в матрицу
mtrx <- rbind(x1, x2, x3, x4, x5, x6, x7, x8)

if (any(mtrx[,1] == c & mtrx[,2] == d)) {
    cat('YES')
} else {
    cat('NO')
}
```

```r
# @title https://stepik.org/lesson/244032/step/14
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
# lines <- c('0', '0') # INF
# lines <- c('6', '-2') # NO
# lines <- c('2', '-8') # 4

a <- as.integer(lines[1])
b <- as.integer(lines[2])

if (a == 0 & b == 0) {
    cat('INF')
} else if ((a == 0 & b != 0) || b %% a != 0) {
    cat('NO')
} else {
    cat(-b / a)
}
```

```r
# @title https://stepik.org/lesson/244032/step/15
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
# lines <- c('0', '0', '2', '2') # INF
# lines <- c('1', '1', '2', '2') # NO
# lines <- c('2', '-8', '1', '1') # 4

a <- as.integer(lines[1])
b <- as.integer(lines[2])
c <- as.integer(lines[3])
d <- as.integer(lines[4])

if (a == 0 & b == 0) {
    cat('INF')
} else if ((a == 0 & b != 0) || b %% a != 0 || (c == 0 & d == 0) || (a / c == b / d)) {
    cat('NO')
} else {
    cat(-b / a)
}
```

```r
# @title https://stepik.org/lesson/244032/step/16
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
# lines <- c('5', '5', '6', '5') # 1 0
# lines <- c('5', '7', '5', '5') # Денег недостаточно

a <- as.integer(lines[1])
b <- as.integer(lines[2])
c <- as.integer(lines[3])
d <- as.integer(lines[4])

if (100 * c + d >= 100 * a + b) {
    cat((100 * c + d - 100 * a - b) %/% 100, (100 * c + d - 100 * a - b) %% 100)
} else {
    cat('Денег недостаточно')
}
```
