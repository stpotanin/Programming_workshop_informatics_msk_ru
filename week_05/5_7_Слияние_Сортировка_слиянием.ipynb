```r
# @title https://stepik.org/lesson/263810/step/1 Задача №1442. Объединение последовательностей
# https://informatics.msk.ru/mod/statements/view.php?chapterid=1442
# https://habr.com/ru/companies/edison/articles/431964/
lines <- readLines(file('stdin'))

# проверка
# lines <- c('2') # 1 4 8
lines <- c('10000')
x <- as.integer(lines[1])

i <- 1
j <- 1
max_val <- x^3
max_i <- floor(max_val^(1/2))
n_max <- max_i + x  # верхняя граница количества уникальных чисел
res <- c()

for (k in 1:n_max) {
    if (i > max_i && j > x) break

    if (i^2 == j^3) {
        res <- c(res, i^2)
        i <- i + 1
        j <- j + 1
    } else if (i <= max_i && (j > x || i^2 < j^3)) {
        res <- c(res, i^2)
        i <- i + 1
    } else {
        res <- c(res, j^3)
        j <- j + 1
    }
}

cat(paste(res, collapse = ' '), '\n')
cat(length(res), '\n')  # 11725

```

```r
# @title https://stepik.org/lesson/263810/step/2 Объединение последовательностей
# https://ru.wikipedia.org/wiki/%D0%A1%D0%BE%D1%80%D1%82%D0%B8%D1%80%D0%BE%D0%B2%D0%BA%D0%B0_%D1%81%D0%BB%D0%B8%D1%8F%D0%BD%D0%B8%D0%B5%D0%BC

merge_sort <- function(arr) {
    if (length(arr) <= 1) return(arr)  # если всё разбили, возвращаем

    # Если длиннее, то делим:
    mid <- length(arr) %/% 2
    left_half <- arr[1:mid]
    right_half <- arr[(mid+1):length(arr)]

    # Рекурсия для каждой половины
    left_half <- merge_sort(left_half)
    right_half <- merge_sort(right_half)

    # После того, как исходный массив рекурсивно разбит на отдельные элементы
    # (которые можно считать отсортированными по определению),
    # начинается процесс их объединения (слияния) в более крупные отсортированные
    # подмассивы, пока не будет восстановлен полностью отсортированный массив.
    return(merge(left_half, right_half))
}

merge <- function(left, right) {
    merged <- c()  # пустой вектор для результата
    # Пока в обеих половинах есть элементы,
    # сравниваем первые из них и добавляем меньший в merged
    while (length(left) > 0 && length(right) > 0) {
        if (left[1] < right[1]) {
            merged <- c(merged, left[1])  # добавляем первый из левой половины
            left <- left[-1]  # удаляем его из левой половины
        } else {
            merged <- c(merged, right[1])  # аналогично для правой половины
            right <- right[-1]
        }
    }
    # Если один из векторов опустел, добавляем оставшиеся элементы другого в конец merged
    merged <- c(merged, left, right)
    return(merged)
}

lines <- readLines(file('stdin'))

# проверка
# lines <- c('5', '5 2 4 1 3') # 1 2 3 4 5

arr <- as.integer(unlist(strsplit(lines[2], ' ')))
sorted_arr <- merge_sort(arr)
cat(sorted_arr)
```

```r
# @title https://stepik.org/lesson/263810/step/2 Задача №767. Похожие массивы
# https://informatics.msk.ru/mod/statements/view.php?chapterid=767
lines <- readLines(file('stdin'))

# проверка
# lines <- c('3', '1 7 9', '4', '9 7 7 1') # YES
# lines <- c('3', '1 7 9', '5', '9 7 7 1 5') # NO

a <- as.integer(unlist(strsplit(lines[2], ' ')))
b <- as.integer(unlist(strsplit(lines[4], ' ')))

count_a <- rep(0, 100000)
count_b <- rep(0, 100000)
for (i in a) count_a[i] <- ifelse(count_a[i] == 0, 1, count_a[i])
for (i in b) count_b[i] <- ifelse(count_b[i] == 0, 1, count_b[i])
count <- count_a + count_b
cat(ifelse(any(count == 1), 'NO', 'YES'))
```
