```r
# @title https://stepik.org/lesson/263808/step/1 Пирамидальная сортировка
# https://campus.epam.kz/ru/blog/462

# Функция heapify для max-кучи
heapify <- function(arr, n, i) {
    largest <- i
    left <- 2 * i
    right <- 2 * i + 1

    if (left <= n && arr[left] > arr[largest]) {
        largest <- left
    }
    if (right <= n && arr[right] > arr[largest]) {
        largest <- right
    }

    if (largest != i) {
        tmp <- arr[i]
        arr[i] <- arr[largest]
        arr[largest] <- tmp
        arr <- heapify(arr, n, largest)
    }
    arr
}

# Функция для построения кучи из массива
build_heap <- function(arr) {
    n <- length(arr)
    for (i in floor(n/2):1) {
        arr <- heapify(arr, n, i)
    }
    arr
}

# Функция для сортировки кучей in-place
heap_sort <- function(arr) {
    n <- length(arr)
    if (n > 1) {
        arr <- build_heap(arr)
        for (i in n:2) {
            tmp <- arr[1]
            arr[1] <- arr[i]
            arr[i] <- tmp
            arr <- heapify(arr, i - 1, 1)
        }
    }
    arr
}

lines <- readLines(file("stdin"))

# для проверки
# lines <- c('5', '5 3 4 2 1') # 1 2 3 4 5
# lines <- c('1', '1') # 1

a <- as.integer(unlist(strsplit(lines[-1], ' ')))
sorted_a <- heap_sort(a)
cat(sorted_a)
```

```r
# @title https://stepik.org/lesson/263808/step/2 Хипуй!
# https://habr.com/ru/articles/263765/
# https://foxford.ru/wiki/informatika/kucha-heap
# https://habr.com/ru/articles/112222/

# Функция heapify для max-кучи
heapify <- function(arr, n, i) {
    largest <- i
    left <- 2 * i
    right <- 2 * i + 1

    if (left <= n && arr[left] > arr[largest]) {
        largest <- left
    }
    if (right <= n && arr[right] > arr[largest]) {
        largest <- right
    }

    if (largest != i) {
        tmp <- arr[i]
        arr[i] <- arr[largest]
        arr[largest] <- tmp
        arr <- heapify(arr, n, largest)
    }
    arr
}

# Функция sift-up (просеивания вверх) для max-кучи
sift_up <- function(arr, i) {
  while (i > 1) {
    parent <- floor(i / 2)
    if (arr[i] > arr[parent]) {
      tmp <- arr[i]
      arr[i] <- arr[parent]
      arr[parent] <- tmp
      i <- parent
    } else {
      break
    }
  }
  arr
}

# Функция для построения кучи из массива
build_heap <- function(arr) {
    n <- length(arr)
    for (i in floor(n/2):1) {
        arr <- heapify(arr, n, i)
    }
    arr
}

# Функция для сортировки кучей in-place
heap_sort <- function(arr) {
    n <- length(arr)
    if (n > 1) {
        arr <- build_heap(arr)
        for (i in n:2) {
            tmp <- arr[1]
            arr[1] <- arr[i]
            arr[i] <- tmp
            arr <- heapify(arr, i - 1, 1)
        }
    }
    arr
}

# добавить в Heap число k
Insert <- function(arr, k) {
    # добавляем в конец
    arr <- c(arr, k)
    # просеиваем вверх при помощи sift_up
    arr <- sift_up(arr, length(arr))
    # теперь максимум в начале
    arr
}

# достать из Heap наибольшее число (удалив его при этом)
Extract <- function(arr) {
    cat(arr[1])
    cat('\n')
    n <- length(arr)
    if (n > 1) {
        arr[1] <- arr[n]
        arr <- arr[-n]
        # здесь хипуем!
        arr <- heapify(arr, length(arr), 1)
        # теперь максимум в начале
    } else {
        arr <- c()
    }
    arr
}



lines <- readLines(file("stdin"))

# для проверки
# lines <- c('2', '0 10000', '1') # 10000
# lines <- c('2', '0 1', '0 2', '0 3', '0 4', '1') # 4
# lines <- c('2', '0 1', '0 2', '0 3', '1', '0 4', '1') # 3 4  в столбик

# Обработка входных команд с вызовом Insert и Extract
arr <- c()
for (line in lines[-1]) {
    command <- as.integer(unlist(strsplit(line, ' ')))
    if (command[1] == 0) {
        arr <- Insert(arr, command[2])
    } else {
        arr <- Extract(arr)
    }
}
```

```r
# @title https://stepik.org/lesson/263808/step/3 Минимум на отрезке TODO

lines <- readLines(file("stdin"))

# для проверки
lines <- c('7 3', '1 3 2 4 5 3 1') # 1 2 2 3 1 в столбик

nk <- as.integer(unlist(strsplit(lines[1], ' ')))
n <- nk[1]
k <- nk[2]
arr <- as.integer(unlist(strsplit(lines[2], ' ')))

window <- arr[1:k]
m <- min(window)
cat(m)
cat('\n')
for (i in (k+1):n) {
    if (arr[i] > m) {
        cat(m)
        cat('\n')
        window <- c(window, arr[i])
    } else {
        m <- arr[i]
        cat(m)
        cat('\n')
        window <- arr[i:k+i]
    }
    cat(window)
    cat('\n')
}
```
