```r
# @title https://stepik.org/lesson/244251/step/1
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
lines <- c('2', '5') # 2 4

a <- as.integer(lines[1])
b <- as.integer(lines[2])

for (i in a:b) {
    if (i %% 2 == 0) cat(i, '')
}
```

```r
# @title https://stepik.org/lesson/244251/step/4
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
# lines <- c('15') # 3

x <- as.integer(lines[1])
res <- x

if (x %% 2 != 0) { # если n нечётное
    for (i in seq(3, x  %/% 2, by = 2)) {
        if (x %% i == 0) {
            res <- i
            break
        }
    }
}   else res <- 2 # если чётное

cat(res)
```

```r
# @title https://stepik.org/lesson/244251/step/5
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
lines <- c('32') # 1 2 4 8 16 32
# lines <- c('1') # 1

x <- as.integer(lines[1])
res <- c()

if (x > 1) {
    for (i in 1:(x %/%2)) {
        if (x %% i == 0) res <- c(res, i)
    }
}

cat(res, x)
```

```r
# @title https://stepik.org/lesson/244251/step/9
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
# lines <- c('3', '1', '2', '3') # 0
# lines <- c('3', '0', '2', '0') # 2

s <- 0
n <- as.numeric(lines[1])
for (i in 2:(n+1)) {
    if (lines[i] == '0') {
        s <- s + 1
    }
}

cat(s)
```

```r
# @title https://stepik.org/lesson/244251/step/10
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
# lines <- c('5', '28', '0', '0' , '0', '0') # 4 1 0

zer <- 0
pos <- 0
n <- as.numeric(lines[1])
for (i in 2:(n+1)) {
    if (lines[i] == '0') {
        zer <- zer + 1
    } else if (as.numeric(lines[i]) > '0') {
        pos <- pos + 1
    }
}

cat(zer, pos, n - zer - pos)
```

```r
# @title https://stepik.org/lesson/244251/step/10 вариант
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
lines <- c('5', '28', '0', '0' , '0', '0') # 4 1 0

n <- as.numeric(lines[1])
vec <- as.numeric(lines[2:(n+1)])
mrtx <- cbind(vec == 0, vec > 0, vec < 0)

cat(colSums(mrtx))
```

```r
# @title https://stepik.org/lesson/244251/step/14
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
# lines <- c('1', '2', '3', '4', '5') # 0

a <- as.numeric(lines[1])
b <- as.numeric(lines[2])
c <- as.numeric(lines[3])
d <- as.numeric(lines[4])
e <- as.numeric(lines[5])
vec <- c()

for (x in 0:1000) if (a * x^3 + b * x^2 + c * x + d == 0 & x != e) vec <- c(vec, x)
cat(length(vec))
```
