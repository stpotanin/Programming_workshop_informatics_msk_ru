```r
# @title https://stepik.org/lesson/244336/step/1
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
# lines <- c('2') # 5

n <- as.integer(lines[1])
k <- 1
r <- 0

while (k <= n) {
    r <- r + k^2
    k <- k + 1
}

cat(r)
```

```r
# @title https://stepik.org/lesson/244336/step/4
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
# lines <- c('2', '1') # 2

n <- as.numeric(lines[1])
k <- as.numeric(lines[2])

factorial <- function (x) {
    f <- 1
    while (x > 0) {
        f <- f * x
        x <- x - 1
    }
    return (f)
}

comb <- function (n, k) {
    return (factorial(n) / factorial(k) / factorial(n - k))
}

cat(comb(n, k))
```
