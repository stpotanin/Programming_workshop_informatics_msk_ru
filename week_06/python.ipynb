```python
# @title https://stepik.org/lesson/464318/step/9 Binary periodical fraction to rational number (perplexity.ai)
import math

s = input().strip()

if '.' in s:
    whole, rest = s.split('.', 1) # знаменатель 1
else:
    whole, rest = s, ""

# Разбираем непериодическую и периодическую часть
if '(' in rest:
    nonrep, rep = rest.split('(', 1)
    rep = rep.rstrip(')')
else:
    nonrep, rep = rest, ""

def b2i(b):
    return int(b, 2) if b else 0

A = b2i(whole)
B = b2i(nonrep)
C = b2i(rep)

lenB = len(nonrep)
lenC = len(rep)

if lenC == 0:
    # без периода
    num = A * (2**lenB) + B
    den = 2**lenB
else:
    # с периодом
    num = A * (2**lenB) * (2**lenC - 1) + B * (2**lenC - 1) + C
    den = (2**lenB) * (2**lenC - 1)

g = math.gcd(num, den)
num //= g
den //= g

print(num, den)

# 10.1(01) # 8 3
# 10.(1) # 3 1
```

```python
# @title https://stepik.org/lesson/464320/step/6 Задача №2798. Возведение в степень
# https://informatics.msk.ru/mod/statements/view.php?chapterid=2798

import time

start = time.perf_counter()

def long_add(x, y):
    # Если длина x меньше длины y, меняем их местами
    if len(x) < len(y):
        x, y = y, x

    # Дописываем нули слева к короткому списку y
    y = [0] * (len(x) - len(y)) + y

    carry = 0
    # Итерация с конца списка к началу (по разрядам)
    for i in range(len(x) - 1, -1, -1):
        s = (x[i] + y[i] + carry) % 10
        carry = (x[i] + y[i] + carry) // 10
        x[i] = s

    # Если есть перенос (carry), добавляем разряд слева
    if carry != 0:
        x = [carry] + x

    return x


def power_of_two(n):
    if n == 0:
        return [1]  # 1 как список цифр

    x = [1]  # 1 как список цифр
    for _ in range(n):
        x = long_add(x, x)
    return x

# print(*power_of_two(int(input())), sep = '')
print(*power_of_two(20000), sep = '')

finish = time.perf_counter()

print(f'Время выполнения: {finish - start:.4f} секунд')
```

```python
# Карацуба:
BASE = 10

def size_in_base(num):
    # Количество цифр в числе num в системе счисления BASE
    return len(str(num))

def split_at(num, d):
    # Разделяем число на старшую часть (hi) и младшую часть (low)
    # # d - количество цифр младшей части
    hi = num // (BASE ** d)
    low = num % (BASE ** d)
    return hi, low

def karatsuba(num1, num2):
    # Базовый случай: если одно из чисел меньше BASE,
    # используем обычное умножение
    if num1 < BASE or num2 < BASE:
        return num1 * num2

    m = max(size_in_base(num1), size_in_base(num2))
    m2 = (m + 1) // 2 # количество цифр младшей части

    high1, low1 = split_at(num1, m2)
    high2, low2 = split_at(num2, m2)

    # Рекурсивные вызовы
    z0 = karatsuba(low1, low2)
    z1 = karatsuba(low1 + high1, low2 + high2)
    z2 = karatsuba(high1, high2)

    # Формула Карацубы: z2*10^(2*m2) + (z1-z2-z0)*10^m2 + z0
    return (z2 * BASE ** (2 * m2)) + ((z1 - z2 - z0) * BASE ** m2) + z0

print(karatsuba(int(input()), int(input())))
```

```python
# строковое представление a и b
def karatsuba_decimal(str_a, str_b):
    """
    Карацуба для десятичных строк. Обходит ограничение Python int (43 цифры).
    Работает с точным количеством цифр.
    """
    # Базовый случай
    if len(str_a) < 10 or len(str_b) < 10:
        return str(int(str_a) * int(str_b))

    m = max(len(str_a), len(str_b))
    m2 = (m + 1) // 2  # ceil(m/2)

    # Точное разделение строк
    low1 = str_a[-m2:] if len(str_a) >= m2 else str_a
    high1 = str_a[:-m2] if len(str_a) > m2 else '0'

    low2 = str_b[-m2:] if len(str_b) >= m2 else str_b
    high2 = str_b[:-m2] if len(str_b) > m2 else '0'

    # Рекурсивные вызовы
    z0 = karatsuba_decimal(low1, low2)
    z1 = karatsuba_decimal(str(int(low1) + int(high1)), str(int(low2) + int(high2)))
    z2 = karatsuba_decimal(high1, high2)

    # Формируем результат
    p1 = z2 + (len(z0) - len(z1) + 2 * m2) * '0'
    p2 = str(int(z1) - int(z2) - int(z0))
    p2 = p2.zfill(m2)  # Дополняем нулями до m2 цифр
    p3 = z0

    result = str(int(p1) + int(p2 + m2 * '0') + int(p3))
    return result.lstrip('0') or '0'

# Универсальная обертка
def karatsuba_big(a, b):
    """Принимает строки или числа, возвращает строку результата"""
    str_a = str(a).lstrip('0') or '0'
    str_b = str(b).lstrip('0') or '0'
    return karatsuba_decimal(str_a, str_b)

# Тест на очень больших числах (>43 цифр)
if __name__ == "__main__":
    # Числа больше 43 цифр
    a = '1' + '23' * 20  # 61 цифра
    b = '4' + '56' * 20  # 61 цифра

    print(f"a ({len(a)} цифр): {a}")
    print(f"b ({len(b)} цифр): {b}")

    result = karatsuba_big(a, b)
    print(f"Результат ({len(result)} цифр): {result}")

    # Проверка на малых числах
    check = karatsuba_big(12345678901234567890, 98765432109876543210)
    print(f"Проверка: {check == str(12345678901234567890 * 98765432109876543210)}")
```

```python
# эвристика Ryhor Abramovich https://stepik.org/lesson/464321/step/2?discussion=11721287&thread=solutions&unit=454967
from math import log, sqrt

n = int(input())
# Проверка 2
if n % 2 == 0:
    print(2, n // 2)
    exit()

# Прямая проверка sqrt(n), ловит квадраты (n = p²)
perfect = int(sqrt(n))
if n % perfect == 0:
    print(perfect, n // perfect)
    exit()

# middle = log2(n)/2 ≈ log2(sqrt(n)), середина логарифмического масштаба
middle = int(log(n, 2)) // 2
i = 3
# j = 2^middle + 1,  кандидат "большого" простого делителя
j = 2 ** middle + 1

# m = perfect/2*2 + 1, нечётный делитель около sqrt(n)/2
# m = (perfect // 2) * 2 + 1
m = perfect | 1

# Тройной поиск: i++ (малые), j++ (большие), m-- (около sqrt)
while i < 2 ** middle:
    if n % i == 0:
        print(min(i, n // i), max(i, n // i))
        exit()

    if n % j == 0:
        print(min(j, n // j), max(j, n // j))
        exit()

    if n % m == 0:
        print(min(m, n // m), max(m, n // m))
        exit()
    i +=2
    j +=2

    m -=2

```

# Алгоритм Полларда ρ (метод факторизации целых чисел)
Алгоритм Полларда ρ — это вероятностный метод факторизации целых чисел, предложенный Джоном Поллардом в 1975 году. Он особенно эффективен для нахождения небольших нетривиальных делителей больших составных чисел и используется как часть многих современных криптоаналитических и вычислительных систем:

```python
import random
import math

def pollard_rho(n):
    if n % 2 == 0:
        return 2
    if n % 3 == 0:
        return 3

    while True:  # 🔥 перезапуск при неудаче
        x = random.randint(2, n - 1)
        y = x
        c = random.randint(1, n - 1)

        def f(v):
            return (v * v + c) % n

        d = 1

        while d == 1:
            x = f(x)
            y = f(f(y))
            d = math.gcd(abs(x - y), n)

        if d != n:   # 🔥 защита от фейла
            return sorted([d, n // d])


print(*pollard_rho(int(input())))


# 274876858369        # M19^2 = 524287 * 524287             # < 1 сек
# 100000627900005643                                        # < 1 сек
# 4611686014132420609 # M31^2 = 2147483647 * 2147483647     # < 1 сек

```
