```r
# @title https://stepik.org/lesson/244343/step/10
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
# lines <- c('179', '20') # -1 :2 -1 :2 :2 -1 -1 в столбик

a <- as.numeric(lines[1])
b <- as.numeric(lines[2])

while (a > b) {
    if (a >= 2 * b & a %% 2 == 0) {
        cat(':2', sep='\n')
        a <- a %/% 2
    } else {
        cat('-1', sep='\n')
        a <- a - 1
    }
}
```

```r
# @title https://stepik.org/lesson/244343/step/11
# https://informatics.msk.ru/mod/statements/view3.php?chapterid=3082
# https://ya-uchitel.ru/load/blogi/metod_reshenija_zadach_dlja_ispolnitelja_vodolej_s_dvumja_sosudami_s_pomoshhju_kratnykh_emkostej/489-1-0-19382

input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
# lines <- c('8', '5', '1') # в столбик
# lines <- c('3', '5', '6') # Impossible
lines <- c('5', '3', '1') # >A A>B >A A>B в столбик

a <- as.numeric(lines[1])
b <- as.numeric(lines[2])
n <- as.numeric(lines[3])

# Решение задачи существует, если количество воды, которое надо отмерить, делится нацело на НОД (А, В)
# и оно не превосходит размер наибольшего сосуда

# функция для НОД рекурсией (abs, т.к в общем виде допустимы отрицательные числа)
gcd <- function(a, b) if (b == 0) return(abs(a)) else return(gcd(b, a %% b))

if (n %% gcd(a, b) == 0 & n <= max(a, b)) { # делится нацело на НОД (А, В) И не превосходит N
    A <- 0
    B <- 0

    for (i in 1:5) {
            # начнем с наполнения сосуда A (А = 8 л)
            if (A == 0) {
                A <- a
                cat('>A', sep='\n')
            }
            # Затем все время переливаем воду в сосуд B (В = 5 л)
            while (A != n & B != n & A != 0)
                # Переливаем сколько можно из A в B без переполнения:
                transfer <- min(A, b - B) # сколько может вместить B или сколько есть в A
                A <- A - transfer
                B <- B + transfer
                cat('A>B\n')
                # Если сосуд B полон, выливаем воду:
                if (A == 0) {
                    A <- a
                    cat('>A\n')
                }
                # Проверяем, достигли ли нужный объем:
                if (A == n | B == n) break

    }
} else cat('Impossible')
```

```r
# @title https://stepik.org/lesson/244343/step/11
# https://informatics.msk.ru/mod/statements/view3.php?chapterid=3082
# https://ya-uchitel.ru/load/blogi/metod_reshenija_zadach_dlja_ispolnitelja_vodolej_s_dvumja_sosudami_s_pomoshhju_kratnykh_emkostej/489-1-0-19382

input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
# lines <- c('8', '5', '1')
# lines <- c('3', '5', '6') # Impossible
# lines <- c('5', '3', '1')
# lines <- c('3', '6', '2') # Impossible
# lines <- c('600', '5', '60')
lines <- c('60', '1', '6')
lines <- c('1', '60', '6')

a <- as.numeric(lines[1])
b <- as.numeric(lines[2])
n <- as.numeric(lines[3])

# Решение задачи существует, если количество воды, которое надо отмерить, делится нацело на НОД (А, В)
# и оно не превосходит размер наибольшего сосуда

# функция для НОД рекурсией (abs, т.к в общем виде допустимы отрицательные числа):
gcd <- function(a, b) if (b == 0) return(abs(a)) else return(gcd(b, a %% b))

if (n %% gcd(a, b) == 0 & n <= max(a, b)) { # делится нацело на НОД (А, В) и не превосходит N
    A <- 0
    B <- 0

    if (a >= b) {
        # Наполняем A в начале
        A <- a
        cat('>A', sep='\n')

        repeat {
            # Проверка достижения цели
            if (A == n) break

            # Переливаем из A в B
            transfer <- min(A, b - B)
            A <- A - transfer
            B <- B + transfer
            cat('A>B', sep='\n')

            # Проверка достижения цели
            if (A == n || B == n) break

            # Если B полный, выливаем
            if (B == b) {
                B <- 0
                cat('B>', sep='\n')
            }

            # Если A пустой, наполняем
            if (A == 0) {
                A <- a
                cat('>A', sep='\n')
            }
        }
    } else {
        # без комментариев
        B <- b
        cat('>B', sep='\n')

        repeat {
            if (B == n) break

            transfer <- min(B, a - A)
            B <- B - transfer
            A <- A + transfer
            cat('B>A', sep='\n')

            if (B == n || A == n) break

            if (A == a) {
                A <- 0
                cat('A>', sep='\n')
            }

            if (B == 0) {
                B <- b
                cat('>B', sep='\n')
            }
        }
    }


} else {
    cat('Impossible')
}
```
