```r
# https://stepik.org/lesson/464321/step/1
isPrime <- function(n) {
    if (n %% 2 == 0) return (n == 2) # True если 2, но False если другое чётное
    d = 3
    # проверяем делимость всех нечётных до
    while (d * d <= n & n %% d != 0) d <- d + 2
    return (d * d > n) # True, если не нашёлся делитель меньше корня из n
}

lines <- readLines(file('stdin'))

# проверка
lines <- c('2') # YES
# lines <- c('3') # YES
# lines <- c('999999999999999989') # YES 83 секунды

prime = isPrime(as.integer(lines[1]))
cat(ifelse(prime, 'YES', 'NO'))

# Failed test #4 of 12. Wrong answer
```

```r
# https://stepik.org/lesson/464321/step/2
library(gmp)

two_divisors <- function(n) {
  if (n %% 2 == 0) return(c(as.bigz(2), n %/% 2))
  if (n %% 3 == 0) return(c(as.bigz(3), n %/% 3))

  d <- as.bigz(5)
  while (d * d <= n) {
    if (n %% d == 0) return(c(d, n %/% d))
    if (n %% (d + 2) == 0) return(c(d + 2, n %/% (d + 2)))
    d <- d + 6
  }

  return(c(n, as.bigz(1)))
}


lines <- readLines(file('stdin'))

# проверка
lines <- c('15') # 3 5
# lines <- c('25') # 3 5
# lines <- c('2147483647') # M31

n <- as.bigz(lines[1])
res <- two_divisors(n)
# cat(res[1], res[2])
cat(as.character(res[1]), as.character(res[2]))
```

```r
# @title Pollard Rho
library(gmp)

gcd <- function(a, b) {
  as.bigz(gcd.bigz(a, b))
}

pollard_rho <- function(n) {
  n <- as.bigz(n)

  if (n %% 2 == 0) return(as.bigz(2))
  if (n %% 3 == 0) return(as.bigz(3))

  repeat {
    x <- as.bigz(sample(2:(as.numeric(n)-1), 1))
    y <- x
    c <- as.bigz(sample(1:(as.numeric(n)-1), 1))

    f <- function(v) (v * v + c) %% n

    d <- as.bigz(1)

    while (d == 1) {
      x <- f(x)
      y <- f(f(y))
      d <- gcd(abs(x - y), n)
    }

    if (d != n) return(d)
  }
}

factor_two <- function(n) {
  n <- as.bigz(n)
  d <- pollard_rho(n)
  return(sort(c(d, n / d)))
}


lines <- readLines(file('stdin'))

# проверка
# lines <- c('15') # 3 5
lines <- c('274876858369')       # 524287 524287
# lines <- c('100000627900005643') # result would be too long a vector (((

n <- as.bigz(lines[1])
res <- factor_two(n)
cat(as.character(res[1]), as.character(res[2]))

# 274876858369        # M19^2 = 524287 * 524287
# 100000627900005643  # result would be too long a vector (((
# 4611686014132420609 # M31^2 = 2147483647 * 2147483647
```
