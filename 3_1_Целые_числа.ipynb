```r
# @title https://stepik.org/lesson/427793/step/1
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
# lines <- c('2 3') # 2

line <- unlist(strsplit(lines, ' '))
n <- as.integer(line[1])
m <- as.integer(line[2])

cat(ifelse(m %% n == 0, m %/% n, m %/% n + 1))
```

```r
# @title https://stepik.org/lesson/427793/step/2 вариант TODO
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
lines <- c('100') # 5050
lines <- c('555555555') # 154320987623456790

n <- as.numeric(lines[1])

if ((n + 1) %% 2 == 0) {
    a <- (n + 1) %/% 2
    b <- n
} else {
    b <- (n + 1) %/% 2
    a <- n
}

# res <- 0
# for (i in 1:b) {
#     res <- res + a
# }

s <- c(unlist(lapply(strsplit(as.character(a), ''), as.numeric)), 0)
lgth <- length(s)

res <- 0
for (i in 1:lgth) {
    res <- res + b * s[i] * 10^(lgth-i-1)
}


cat(format(res, scientific = FALSE))

# 555555555 * 78
```

```r
# @title https://stepik.org/lesson/427793/step/3
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
# lines <- c('13257') # It is 3 hours 40 minutes.

k <- as.integer(lines[1])

h <- k %/% 3600
m <- k %% 3600 %/% 60

cat(sprintf('It is %d hours %d minutes.', h, m))
```

```r
# @title https://stepik.org/lesson/427793/step/4
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
# lines <- c('105') # It is 3 hours 30 minutes.

k <- as.integer(lines[1])

h <- k %/% 30
m <- k %% 30 * 2

cat(sprintf('It is %d hours %d minutes.', h, m))
```

```r
# @title https://stepik.org/lesson/427793/step/5
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
# lines <- c('50 1') # 1 1
# lines <- c('50 50') # 1 50
# lines <- c('50 149') # 3 49

line <- unlist(strsplit(lines, ' '))

k <- as.integer(line[1])
n <- as.integer(line[2])

p <- (n-0.5) %/% k + 1
i <- n - (p-1) * k

cat(p, i)
```

```r
# @title https://stepik.org/lesson/427793/step/6
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
# lines <- c('1 1 1', '2 2 2') # 3661

t1 <- as.integer(unlist(strsplit(lines[1], ' ')))
t2 <- as.integer(unlist(strsplit(lines[2], ' ')))

x1 <- t1[1] * 3600 + t1[2] * 60 + t1[3]
x2 <- t2[1] * 3600 + t2[2] * 60 + t2[3]

cat(x2 - x1)
```
