В языке R есть аналоги побитовых операций, но с некоторыми особенностями:

- Побитовое И (`&` в Python) в R реализуется оператором bitwAnd(x, y).
- Побитовое ИЛИ (`|` в Python) реализуется через bitwOr(x, y).
- Побитовое исключающее ИЛИ (`^` в Python) — это bitwXor(x, y).
- Побитовое НЕ (`~` в Python) — в R нет прямого побитового отрицания как у Python, но его можно реализовать через bitwNot(x).
- Сдвиги влево (`<<`) и вправо (`>>`) в R выполняются с помощью bitwShiftL(x, n) и bitwShiftR(x, n), где n — количество сдвигаемых бит.

Таким образом, в R побитовые операции реализуются через встроенные функции из базового пакета, которые применяются к целочисленным значениям, например:
```r
bitwAnd(5, 3)    # аналог 5 & 3
bitwOr(5, 3)     # аналог 5 | 3
bitwXor(5, 3)    # аналог 5 ^ 3
bitwNot(5)       # аналог ~5
bitwShiftL(5, 2) # аналог 5 << 2
bitwShiftR(5, 1) # аналог 5 >> 1
```

````r
# @title https://stepik.org/lesson/464319/step/1 2^n
lines <- readLines(file('stdin'))

# проверка
# lines <- c('2') # 2
# lines <- c('30') # 1073741824
# lines <- c('31') # 2147483648

n <- as.numeric(lines[1])
cat(ifelse(n < 31, bitwShiftL(1, n), 2^n))
````

````r
# @title https://stepik.org/lesson/464319/step/2 2^n+2^m
bitwise_add <- function(a, b) {
    while (b != 0) {
        carry = bitwAnd(a, b)    # вычисляем перенос a & b
        a = bitwXor(a, b)        # сумма без переноса XOR
        b = bitwShiftL(carry, 1) # перенос сдвигается влево на один бит carry << 1
    }
    a
}

lines <- readLines(file('stdin'))

# проверка
# lines <- c('1 2') # 6
# lines <- c('8 8') # 512

a <- as.integer(strsplit(lines[1], ' ')[[1]])

x <- ifelse(a[1] < 31, bitwShiftL(1, a[1]), 2^a[1])
y <- ifelse(a[2] < 31, bitwShiftL(1, a[2]), 2^a[2])

cat(ifelse(x < 31 & y < 31, bitwise_add(x, y), x + y)) # x | (2 XOR y)
````

````r
# @title https://stepik.org/lesson/464319/step/2 Задача №122. 2^n+2^m
# https://informatics.msk.ru/mod/statements/view.php?chapterid=122
bitwise_add <- function(a, b) {
    while (b != 0) {
        carry = bitwAnd(a, b)    # вычисляем перенос a & b
        a = bitwXor(a, b)        # сумма без переноса XOR
        b = bitwShiftL(carry, 1) # перенос сдвигается влево на один бит carry << 1
    }
    a
}

lines <- readLines(file('stdin'))

# проверка
# lines <- c('1 2') # 6
# lines <- c('31 31') # 4294967296

a <- as.integer(strsplit(lines[1], ' ')[[1]])

# с защитой от >= 31
use_safe <- max(a) < 31
if (use_safe) {
    x <- bitwShiftL(1, a[1])
    y <- bitwShiftL(1, a[2])
    cat(bitwise_add(x, y))
} else {
    x <- 2^a[1]
    y <- 2^a[2]
    cat(x + y)
}

# # без защиты (> 31)
# x <- bitwShiftL(1, a[1])
# y <- bitwShiftL(1, a[2])
# cat(bitwise_add(x, y))
````

````r
# @title https://stepik.org/lesson/464319/step/3 Установить значение бита в 1
lines <- readLines(file('stdin'))

# проверка
# lines <- c('5 1') # 7
# lines <- c('30 30') # 1073741854
# lines <- c('33 33') # 8589934625

a <- as.numeric(strsplit(lines[1], ' ')[[1]])
if (a[2] < 32) {
    cat(bitwOr(a[1], bitwShiftL(1, a[2]))) # по условию
} else {
    bits <- as.integer(intToBits(a[1])) # при i > 30
    bits[32:1000] <- 0
    bits[a[2] + 1] <- 1
    decimal <- sum(rev(bits) * 2^((length(bits)-1):0))
    cat(decimal)
}
````

````r
# @title https://stepik.org/lesson/464319/step/4 Установить значение бита в 0
lines <- readLines(file('stdin'))

# проверка
# lines <- c('31 2') # 27
# lines <- c('32 2') # 32

a <- as.integer(strsplit(lines[1], ' ')[[1]])
cat(bitwAnd(a[1], bitwNot(bitwShiftL(1, a[2]))))
````

````r
# @title https://stepik.org/lesson/464319/step/5 Определить значение бита
dec_to_bin <- function(x) {
  if (x == 0) "0" else if (x == 1) "1" else paste0(dec_to_bin(x %/% 2), x %% 2)
}

lines <- readLines(file('stdin'))

# проверка
# lines <- c('2147483647 1') # 1
# lines <- c('2147483648 1') # 0

a <- as.numeric(strsplit(lines[1], ' ')[[1]])

if (-2147483648 <= a[1] & a[1] <= 2147483647) {
    # по условию
    cat(bitwAnd(bitwShiftR(a[1], a[2]), 1))
} else {
    # при выходе за диапазон int32
    b <- dec_to_bin(a[1])
    b_list <- strsplit(b, '')[[1]]
    cat(rev(b_list)[a[2]])
}
````

````r
# @title https://stepik.org/lesson/464319/step/6 Инвертировать бит
lines <- readLines(file('stdin'))

# проверка
# lines <- c('5 2') # 1

a <- as.numeric(strsplit(lines[1], ' ')[[1]])

cat(bitwXor(a[1], bitwShiftL(1, a[2])))
````

````r
# @title https://stepik.org/lesson/464319/step/8 Обнулить последние биты
# Кавалеры приглашают дамов
lines <- readLines(file('stdin'))

# проверка
# lines <- c('5 1') # 4

a <- as.numeric(strsplit(lines[1], ' ')[[1]])

cat(bitwShiftL(bitwShiftR(a[1], a[2]), a[2]))
````

````r
# @title https://stepik.org/lesson/464319/step/9 Обнулить все биты, кроме последних
lines <- readLines(file('stdin'))

# проверка
# lines <- c('5 1') # 1
# lines <- c('12345 6') # 57
# lines <- c('12345 5') # 25

a <- as.numeric(strsplit(lines[1], ' ')[[1]])

# a & ~(~0 << i)
cat(bitwAnd(a[1], bitwNot(bitwShiftL(bitwNot(0), a[2]))))

# # для проверки
# cat('\n')
# cat(bitwShiftR(bitwShiftL(a[1], 32 - a[2]), 32 - a[2]))
````

````r
# @title https://stepik.org/lesson/464319/step/10 Двоичные строки заданной длины
# функция для перевода из десятичной в двоичную (строку длины n)
dec_to_bin_bitwise <- function(x, n) {
    y <- ''
    for (i in 0:(n-1)) {  # i от 0 до n-1 для старших битов
        bit <- c('0', '1')[(bitwAnd(bitwShiftL(1, n-1-i), x) > 0) + 1]
        y <- paste0(y, bit)
    }
    y
}

lines <- readLines(file('stdin'))

# проверка
# lines <- c('2') # 00 01 10 11 в столбик

n <- as.numeric(lines[1])

# все двоичные числа в нужном диапазоее
for (x in 0:(bitwShiftL(1, n)-1)) {
    cat(dec_to_bin_bitwise(x, n), sep = '\n')
}
````
