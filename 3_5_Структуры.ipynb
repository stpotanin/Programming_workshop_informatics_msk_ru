````r
# @title https://stepik.org/lesson/427796/step/2
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
# lines <- c('2', '1 2', '2 3') # 1.5 2.5
# lines <- c('2', '1 1', '1 3') # # 1.0 2.0
# lines <- c('3', '1 2', '2 3', '2 33') # 1.6666666666 12.6666666666

# функция для обрезания 10 знаков
truncate <- function(x, digits = 10) {
  s <- format(x, scientific = FALSE, digits = digits + 3)
  if (grepl("\\.", s)) {
    parts <- strsplit(s, "\\.")[[1]]
    integer_part <- parts[1]
    decimal_part <- substr(parts[2], 1, digits)
    return(paste0(integer_part, ".", decimal_part))
  } else {
    return(s)
  }
}

s <- lines[2:length(lines)]

d = (sapply(s, function(x) as.numeric(unlist(strsplit(x, ' ')))))

m <- as.numeric(rowMeans(d))

formatted_m <- sapply(m, function(x) {
  if (floor(x) == x) {
    sprintf("%.1f", x)
  } else {
    truncate(x)
  }
})

cat(formatted_m)
````

````r
# @title https://stepik.org/lesson/427796/step/5
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
# lines <- c('2', '0 0', '0 1', '1 0', '1 1') # 3.414213562373
# lines <- c('6', '101 44', '54 38', '93 75', '45 21', '100 37', '55 99') # 212.527825873008

s <- lines[-1]
triangles <- combn(s, 3, byrow=FALSE)

perimeters <- apply(triangles, 2, function(triangle_vertices) {
    points <- lapply(triangle_vertices, function(vertex) as.numeric(unlist(strsplit(vertex, ' '))))

    dist12 <- sqrt((points[[2]][1] - points[[1]][1])^2 + (points[[2]][2] - points[[1]][2])^2)
    dist13 <- sqrt((points[[3]][1] - points[[1]][1])^2 + (points[[3]][2] - points[[1]][2])^2)
    dist23 <- sqrt((points[[3]][1] - points[[2]][1])^2 + (points[[3]][2] - points[[2]][2])^2)

  return(sum(dist12, dist13, dist23))
})

m <- max(perimeters)

# функция для обрезания 12 знаков
truncate <- function(x, digits = 12) {
  s <- format(x, scientific = FALSE, digits = digits + 5)
  if (grepl("\\.", s)) {
    parts <- strsplit(s, "\\.")[[1]]
    integer_part <- parts[1]
    decimal_part <- substr(parts[2], 1, digits)
    return(paste0(integer_part, ".", decimal_part))
  } else {
    return(s)
  }
}

# функция для выбора форматирования
formatted_m <- sapply(m, function(x) {
  if (floor(x) == x) {
    sprintf("%.1f", x)
  } else {
    truncate(x)
  }
})

cat(formatted_m)
````

````r
# @title https://stepik.org/lesson/427796/step/6
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
# lines <- c('4', '0 0', '0 1', '1 0', '1 1') # 0.5
# lines <- c('6', '101 44', '54 38', '93 75', '45 21', '100 37', '55 99') # 2069.0

s <- lines[-1]
triangles <- combn(s, 3, byrow=FALSE)

area <- apply(triangles, 2, function(triangle_vertices) {
    points <- lapply(triangle_vertices, function(vertex) as.numeric(unlist(strsplit(vertex, ' '))))

    a <- sqrt((points[[2]][1] - points[[1]][1])^2 + (points[[2]][2] - points[[1]][2])^2)
    b <- sqrt((points[[3]][1] - points[[1]][1])^2 + (points[[3]][2] - points[[1]][2])^2)
    c <- sqrt((points[[3]][1] - points[[2]][1])^2 + (points[[3]][2] - points[[2]][2])^2)
    p <- (a + b + c) / 2

  return(sqrt(p * (p - a) * (p -b) * (p - c)))
})

max_area <- max(area)

# функция для обрезания 2 знаков
truncate <- function(x, digits = 2) {
  s <- format(x, scientific = FALSE, digits = digits + 5)
  if (grepl("\\.", s)) {
    parts <- strsplit(s, "\\.")[[1]]
    integer_part <- parts[1]
    decimal_part <- substr(parts[2], 1, digits)
    return(paste0(integer_part, ".", decimal_part))
  } else {
    return(s)
  }
}

# функция проверки числа на целое с допуском
is_integer <- function(x, tol = 1e-8) {
  abs(x - round(x)) < tol
}

# форматирование результата с проверкой на целое число
if (is_integer(max_area)) {
  formatted_max_area <- sprintf("%.1f", max_area)
} else {
  formatted_max_area <- truncate(max_area)
}

cat(formatted_max_area)
````

````r
# @title https://stepik.org/lesson/427796/step/7
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
# lines <- c('2', 'Markov Valeriy 4 5 2', 'Kozlov Georgiy 5 1 2') # 4.5 3.0 2.0
# lines <- c("10", "t t 2 3 3", "t t 2 2 4", "t t 3 5 3", "t t 3 5 1", "t t 1 2 1",
#                  "t t 2 5 1", "t t 3 2 1", "t t 5 5 5", "t t 3 5 1", "t t 3 5 4") # 2.7 3.9 2.4

s <- lines[-1]
data <- strsplit(s, ' ')
grades <- sapply(data, function(x) as.numeric(x[-c(1, 2)]))
m <- rowMeans(grades)

# функция для обрезания 15 знаков
truncate <- function(x, digits = 15) {
  # округление с нужной точностью
  x_rounded <- round(x, digits)
  s <- format(x_rounded, scientific = FALSE, trim = TRUE, nsmall = digits)
  if (grepl("\\.", s)) {
    parts <- strsplit(s, "\\.")[[1]]
    integer_part <- parts[1]
    decimal_part <- substr(parts[2], 1, digits)
    # убираем лишние нули справа
    decimal_part <- sub("0+$", "", decimal_part)
    if (nchar(decimal_part) == 0) {
      return(integer_part)
    } else {
      return(paste0(integer_part, ".", decimal_part))
    }
  } else {
    return(s)
  }
}

# функция проверки числа на целое с допуском
is_integer <- function(x, tol = 1e-8) {
  abs(x - round(x)) < tol
}

# функция для выбора форматирования
formatted_m <- function(x) if (is_integer(x)) {
    sprintf("%.1f", x)
    } else {
        truncate(x)
    }

cat(sapply(m, formatted_m))
````

````r
# @title https://stepik.org/lesson/427796/step/12
input <- file("stdin")
lines <- readLines(input, warn=FALSE)
if ((length(lines) - 1) %% 4 != 0) {
  lines <- c(lines, "")  # добавляем пустую строку если её забыли
}
close(input)

# для проверки, должно "развернуть"
# lines <- c('1',
#            'Vishnya', 'Ostap', '11D', '20.07.93',
#            'Zadornov', 'Mark', '11A','20.07.93',
#            'Sidorov', 'Sidor', '9A', '20.07.93')

s <- lines[-1]
mtrx <- matrix(s, ncol=4, byrow=TRUE)

# Функция для разделения класса на число и букву
class_rank <- function(x) {
  chars <- strsplit(x, '')
  # числовая часть
  nums <- sapply(chars, function(ch) as.integer(paste(ch[-length(ch)], collapse='')))
  # буква
  lets <- sapply(chars, function(ch) ch[length(ch)])
  list(nums = nums, lets = lets) # для обращения к результату по именам
}

rank <- class_rank(mtrx[,3]) # сразу для всех строк
# Сортируем по числу, по букве, по фамилии
order_index <- order(rank$nums, rank$lets, mtrx[,1])
m <- mtrx[order_index, , drop=FALSE]

for (i in 1:nrow(m)) {
  cat(m[i,3], ' ', m[i,1], ' ', m[i,2], ' ', m[i,4], '\n', sep='')
}
````
