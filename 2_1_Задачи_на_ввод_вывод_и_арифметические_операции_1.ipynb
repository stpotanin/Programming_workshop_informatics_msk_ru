```r
# @title https://stepik.org/lesson/244030/step/1 Гипотенуза

input <- file("stdin")
lines <- readLines(input)
close(input)

# Теперь lines — это вектор строк, каждая строка — одна строка ввода

a <- as.integer(lines[1])
b <- as.integer(lines[2])

c <- round(sqrt(a^2 + b^2))
cat(c)
```

```r
# @title https://stepik.org/lesson/244030/step/2 Следующее и предыдущее
input <- file("stdin")
lines <- readLines(input)
close(input)

a <- as.integer(lines[1])

# a = 179 # для проверки

# Использование sprintf для форматирования строки в R
cat(sprintf('The next number for the number %d is %d.\n', a, a+1))
# Использование paste0 для объединения строк в R
cat(paste0('The previous number for the number ', a, ' is ', a-1, '.'))
```

```r
# @title https://stepik.org/lesson/244030/step/9 Сумма цифр
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
lines <- c('-192') # 12

n <- abs(as.integer(lines[1]))
tokens <- strsplit(as.character(n), '')[[1]]
digits <- unlist(lapply(tokens, function(x) as.integer(x)))
cat(sum(digits))
```

```r
# @title https://stepik.org/lesson/244030/step/9 Сумма цифр (вариант)
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
# lines <- c('-192') # 12

n <- abs(as.integer(lines[1]))
s <- 0
while (n > 0) {
  s <- s + n %% 10
  n <- n %/% 10
}
cat(s)
```

```r
# @title https://stepik.org/lesson/244030/step/11 Электронные часы - 1
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
# lines <- c('150') # 2 30

n <- as.integer(lines[1])

hours <- n %% 1440 %/% 60
min <- n %% 60
cat(hours, min)
```

```r
# @title https://stepik.org/lesson/244030/step/12 Электронные часы - 2
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
# lines <- c('3602') # 1:00:02
# lines <- c('12345') # 3:25:45

n <- as.integer(lines[1])

hours <- n %% 86400 %/% 3600
min <- n %% 86400 %% 3600 %/% 60
sec <- n %% 60

cat(sprintf("%d:%02d:%02d", hours, min, sec))
```

```r
# @title https://stepik.org/lesson/244030/step/13 Обмен значений
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
# lines <- c('8', '5') # 5 8
# lines <- c('11', '19') # 19 11

a <- as.integer(lines[1])
b <- as.integer(lines[2])

a <- a + b
b <- a - b
a <- a - b

cat(a, b)
```

```r
# @title https://stepik.org/lesson/244030/step/14 Конец уроков
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
# lines <- c('3') # 11 35

lesson <- as.integer(lines[1])

begin <- 9 * 60
end <- begin + lesson * 45 + (lesson - 1) * 15 - lesson %/% 2 * 10

hours <- end %/% 60
min <- end %% 60

cat(hours, min)
```

```r
# @title https://stepik.org/lesson/244030/step/15 Стоимость покупки
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
lines <- c('10', '15', '2') # 20 30

a <- as.integer(lines[1])
b <- as.integer(lines[2])
n <- as.integer(lines[3])

total <- (a * 100 + b) * n

cat(total %/% 100, total %% 100)
```

```r
# @title https://stepik.org/lesson/244030/step/16 Разность времен
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
# lines <- c('1',  '1',  '1',  '2',  '2', '2') # 3661

h1 <- as.integer(lines[1])
m1 <- as.integer(lines[2])
s1 <- as.integer(lines[3])
h2 <- as.integer(lines[4])
m2 <- as.integer(lines[5])
s2 <- as.integer(lines[6])

cat(h2 * 3600 + m2 * 60 + s2 - h1 * 3600 - m1 * 60 - s1)
```
