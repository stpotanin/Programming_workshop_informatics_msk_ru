```r
# @title https://stepik.org/lesson/244046/step/1 Линейный поиск - 1
lines <- readLines(file("stdin"))

# для проверки
# lines <- c('5', '1 2 3 4 5', '3') # 1

a <- unlist(strsplit(lines[2], ' '))
x <- lines[3]
cat(sum(a == x))
```

```r
# @title https://stepik.org/lesson/244046/step/2 Линейный поиск - 2
lines <- readLines(file("stdin"))

# для проверки
# lines <- c('5', '1 2 3 4 5', '3') # YES

a <- unlist(strsplit(lines[2], ' '))
x <- lines[3]
cat(ifelse((any(a == x)), 'YES', 'NO'))
```

```r
# @title https://stepik.org/lesson/244046/step/3 Ближайшее число
lines <- readLines(file("stdin"))

# для проверки
# lines <- c('5', '1 2 3 4 5', '6') # 5

a <- as.integer(unlist(strsplit(lines[2], ' ')))
x <- as.integer(lines[3])

i <- abs(a - x)
cat(a[i == min(i)])
```

```r
# @title https://stepik.org/lesson/244046/step/4 Линейный поиск - 3
lines <- readLines(file("stdin"))

# для проверки
# lines <- c('5', '1 2 3 4 3', '3') # 3 5

a <- as.integer(unlist(strsplit(lines[2], ' ')))
x <- as.integer(lines[3])
i <- 1:length(a)
cat(i[a == x])
```

```r
# @title https://stepik.org/lesson/244046/step/5 Максимальный элемент массива
lines <- readLines(file("stdin"))

# для проверки
lines <- c('5', '1 2 3 4 5') # 5

a <- as.integer(unlist(strsplit(lines[2], ' ')))

m <- -Inf
for (i in 1:length(a)) {
    if (a[i] >= m) m <- a[i]
}

cat(m)
```

```r
# @title https://stepik.org/lesson/244046/step/6 Номер максимального элемента массива
lines <- readLines(file("stdin"))

# для проверки
# lines <- c('5', '5 4 3 2 1') # 1

a <- as.integer(unlist(strsplit(lines[2], ' ')))

i <- 1
m <- -Inf
for (i in 1:length(a)) {
    if (a[i] >= m) {
        m <- a[i]
        max_i <- i
    }
}

cat(max_i)
```

```r
# @title https://stepik.org/lesson/244046/step/8 Столбцы
lines <- readLines(file("stdin"))

# для проверки
# lines <- c('1789', '2', '1789 1111', '1111 1111') # YES NO в столбик
lines <- c('9', '3', '1 2 9', '3 4 5', '6 7 8') # NO NO YES

x <- as.integer(lines[1])
n <- as.integer(lines[2])
a <- matrix(as.integer(unlist(sapply(lines[-c(1, 2)],
                                     function(x) strsplit(x, ' ')))),
            nrow = n,
            byrow = TRUE)

for (i in 1:ncol(a)) cat(ifelse((x %in% a[, i]), 'YES', 'NO'), sep='', end='\n')
```

```r
# @title https://stepik.org/lesson/244046/step/9 Матрица
lines <- readLines(file("stdin"))

# для проверки
# lines <- c('2 2', '0 0', '0 0') # 4
lines <- c('3 4', '3 5 6 7', '4 5 3 9', '5 6 8 7') # 1

a <- matrix(as.integer(unlist(sapply(lines[-1],
                                     function(x) strsplit(x, ' ')))),
            nrow = length(lines) - 1,
            byrow = TRUE)

count <- 0
for (i in 1:nrow(a)) {
    for (j in 1:ncol(a)) {
        if (a[i, j] == min(a[i, ]) & a[i, j] == max(a[, j]))
            count <- count + 1
    }
}

cat(count)
```

```r
# @title https://stepik.org/lesson/244046/step/9 Матрица (вариант)
lines <- readLines(file("stdin"))

# для проверки
# lines <- c('2 2', '0 0', '0 0') # 4
lines <- c('3 4', '3 5 6 7', '4 5 3 9', '5 6 8 7') # 1

a <- matrix(as.integer(unlist(strsplit(lines[-1], ' '))),
            nrow = length(lines) - 1, byrow = TRUE)

min_values <- apply(a, 1, min) # 1 - по строкам
max_values <- apply(a, 2, max) # 2 - по столбцам

is_min <- a == matrix(min_values, nrow = nrow(a), ncol = ncol(a), byrow = FALSE)
is_max <- a == matrix(max_values, nrow = nrow(a), ncol = ncol(a), byrow = TRUE)

cat(sum(is_min & is_max))
```
