```r
# @title https://stepik.org/lesson/464316/step/1 Сложить две дроби

# Функция для НОД
gcd <- function(a, b) ifelse(b == 0, a,  gcd(b, a %% b))
# проверка функции
# gcd(15102, 2052) # 18

lines <- readLines(file('stdin'))

# проверка
# lines <- c('1 3 1 2') # 5 6

x <- as.integer(unlist(strsplit(lines[1], ' ')))
numerator <- x[1] * x[4] + x[2] * x[3]
denominator <- x[2] * x[4]

g <- gcd(numerator, denominator)

numerator <- numerator %/% g
denominator <- denominator %/% g

cat(numerator, denominator)
```

```r
# @title https://stepik.org/lesson/464316/step/2 Алгоритм Евклида

# Функция для НОД вычитанием (Алгоритм Эвклида)
gcd_sub <- function(a, b) ifelse(a == b, b,  ifelse(a < b, gcd_sub(b, b - a), gcd_sub(b, a - b)))
# проверка функции
# gcd_sub(15102, 2052) # 18

lines <- readLines(file('stdin'))

# проверка
# lines <- c('12', '14') # 2

n <- as.integer(lines[1])
m <- as.integer(lines[2])

cat(gcd_sub(n, m))
```

```r
# @title https://stepik.org/lesson/464316/step/3 Теорема Лагранжа
lagrange <- function(n) {
    # Если ноль
    if (n == 0) return(0)

    # Предельное основание квадрата
    max_val <- floor(sqrt(n))
    squares <- (0:max_val)^2

    # Проверки
    # Один вкадрат
    if (n %in% squares) return(c(sqrt(n)))

    # Два квадрата
    for (i in 1:length(squares)) {
        if ((n - squares[i]) %in% squares) {
        return(c(i-1, sqrt(n - squares[i])))
        }
    }

    # Три квадрата
    for (i in 1:length(squares)) {
        r1 <- n - squares[i]
        for (j in 1:length(squares)) {
        if (squares[j] > r1) break # слишком большие
        if ((r1 - squares[j]) %in% squares) {
            return(c(i-1, j-1, sqrt(r1 - squares[j])))
        }
        }
    }

  # Четыре квадрата
  for (i in 1:length(squares)) {
        r1 <- n - squares[i]
        for (j in 1:length(squares)) {
        if (squares[j] > r1) break  # слишком большие
        r2 <- r1 - squares[j]
        for (k in 1:length(squares)) {
            if (squares[k] > r2) break  # слишком большие
            if ((r2 - squares[k]) %in% squares) {
            return(c(i-1, j-1, k-1, sqrt(r2 - squares[k])))
            }
        }
        }
    }
}

lines <- readLines(file('stdin'))

# проверка
# lines <- c('3') # 1 1 1
# lines <- c('167') # 1 2 9 9

n <- as.integer(lines[1])

cat(lagrange(n))
```

```r
# @title https://stepik.org/lesson/464316/step/5 Дружественные числа

divisors <- function(n) {
  divs <- c()
  if (n <= 0) return(divs) # пусто на случай 0 и меньше
  for (i in 1:(n%/%2 + 1)) {
    if (n %% i == 0) {
      divs <- c(divs, i)
    }
  }
  return(divs)
}

lines <- readLines(file('stdin'))

# проверка
# lines <- c('300') # 220 284
# lines <- c('6300') # 220 284

k <- as.integer(lines[1])

for (m in 1:k) {
    n <- sum(divisors(m))
    if (m == sum(divisors(n)) && m < n && n < k) {
        cat(m, n)
        cat('\n')
    }
}
```

```r
# @title https://stepik.org/lesson/464316/step/6 Гипотеза Гольдбаха

is_prime <- function(n) {
    if (n %% 2 == 0) return (n == 2) # True если 2, но False если другое чётное
    d = 3
    # проверяем делимость
    while (d * d <= n && n %% d != 0) {
        d <- d + 2    # всех нечётных до
    }
    return (d * d > n) # True, если не нашёлся делитель меньше корня из n
}

lines <- readLines(file('stdin'))

# проверка
# lines <- c('4') # 2 2
# lines <- c('6') # 3 3
# lines <- c('188') # 61 127


n <- as.integer(lines[1])

a <- 2:n
prime_list <- a[sapply(a, is_prime)]
for (i in prime_list) {
    if ((n - i) %in% prime_list) {
        cat(i, n - i)
        break
    }
}
```

```r
# @title https://stepik.org/lesson/464316/step/9 Полярные единички (медленный вариант)
lines <- readLines(file('stdin'))

# проверка
lines <- c('100') # NO
lines <- c('57') # 18 едиинчек
lines <- c('455261') # 91052 единичек



n <- as.integer(lines[1])
res <- 'NO'

# Функция вычисления остатка от деления очень большого числа (представленного строкой)
# на divisor, не преобразуя всю строку в число (что невозможно для очень длинных чисел)

mod_check <- function(num_str, divisor) {
  rem <- 0  # переменная для хранения текущего остатка, изначально равна 0
  digits <- strsplit(num_str, NULL)[[1]]  # разбиваем строку на отдельные символы (цифры)

  # Проходим по каждой цифре в числовой строке
  for (d in digits) {
    # Преобразуем текущий символ в число
    digit <- as.integer(d)

    # Обновляем остаток: умножаем текущий остаток на 10, прибавляем текущую цифру,
    # и берем по модулю divisor, чтобы сохранить остаток деления
    rem <- (rem * 10 + digit) %% divisor
  }

  # Возвращаем итоговый остаток после обработки всех цифр
  return(rem)
}


i <- 1
while (i < 10^200) {
  num_str <- paste(rep('1', i), collapse = '')
  if (mod_check(num_str, n) == 0) {
    res <- num_str
    break
  }
  i <- i + 1
}

cat(res)
```

```r
# @title https://stepik.org/lesson/464316/step/9 Полярные единички (быстрый вариант)
lines <- readLines(file('stdin'))

# проверка
# lines <- c('100') # NO
# lines <- c('1') # 1
# lines <- c('57') # 18 едиинчек
# lines <- c('455261') # 91052 единичек



n <- as.integer(lines[1])
res <- ifelse(n == 1, 1, 'NO')

r <- 10 %% n
for (i in 1:(10^6 + 1)) {
    if (r == 1) {
        res = rep('1',  i)
        break
    }
    r <- (r * 10) %% n
}

cat(res, sep='')
```

```r
# @title https://stepik.org/lesson/464316/step/10 Разложение на простые++
lines <- readLines(file('stdin'))

# проверка
lines <- c('511') # 2
# lines <- c('1008') # 2^4*3^2*7

n <- as.integer(lines[1])

pd_list <- function(n) {
    pdl <- c()
    d = 2
    while (d * d <= n) {
        if (n %% d == 0 ) {  # находим множитель
            pdl <- c(pdl, d) # добавляем в список
            n <- n %/% d     # кратное делаем делимым
        } else {
            d <- d + 1 # если больше не делится на этот делитель
        }
    }
    if (n > 1) pdl <- c(pdl, n) # если ни на что не поделилось
    return (pdl)
    }

pdl <- pd_list(n)

res <- ''
for (d in unique(pdl)) {
    i <- length(pdl[pdl == d]) # степень множителя
    # рисуем формулу
    if (i > 1) res <- paste(res, d, '^', i, '*', sep='')
    else res <- paste(res, d, '*', sep='')
}

res <- substr(res, 1, nchar(res) - 1) # обрезаем последний *
cat(res)
```

```r
# @title https://stepik.org/lesson/464316/step/11 Разрезание на квадраты
lines <- readLines(file('stdin'))

# проверка
lines <- c('1000000000 1') # 1
# lines <- c('15 3') # 5
# lines <- c('12 8') # 3
# lines <- c('5 5') # 1
# lines <- c('92 777') # 24

n <- as.integer(unlist(strsplit(lines[1], ' ')))
a <- n[1]
b <- n[2]

i <- 0
while (b) {
    i <- i + a %/% b
    r = a %% b
    a <- b
    b <- r
}

cat(format(i, scientific = FALSE))
```
