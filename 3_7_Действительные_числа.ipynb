```r
# @title https://stepik.org/lesson/427798/step/1
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
# lines <- c('2', '3', '7') # NO
# lines <- c('0.1', '0.2', '0.3') # YES

d <- lapply(lines, as.numeric)

if (abs(d[[1]] + d[[2]] - d[[3]]) < 1e-8) cat('YES') else cat('NO')
```

```r
# @title https://stepik.org/lesson/427798/step/2
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
# lines <- c('10 30') # 4

s = as.numeric(unlist(strsplit(lines[1], ' ')))

count <- 1
while (s[[1]] < s[[2]]) {
    s[[1]] <- s[[1]] * 1.7
    count <- count + 1
}

cat(count)
```

```r
# @title https://stepik.org/lesson/427798/step/3
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
# lines <- c('1 20') # 6

s = as.numeric(unlist(strsplit(lines[1], ' ')))

count <- 1
total_dist <- s[[1]]
while (total_dist < s[[2]]) {
    s[[1]] <- s[[1]] * 1.7
    total_dist <- total_dist + s[[1]]
    count <- count + 1
}

cat(count)
```

```r
# @title https://stepik.org/lesson/427798/step/4
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
# lines <- c('1.0 1.0 1.0', '3', '1 0 0 1', '0 0.5 0 2', '0 0 0.25 4') # YES
# lines <- c('1.0 1.0 1.0', '3', '0.5 0 0 1', '0.24 0.5 0 2', '0 0 0.25 4') # NO

norm <- as.numeric(unlist(strsplit(lines[1], ' ')))
total <- c(0, 0, 0)
for (line in lines[-c(1, 2)]) {
    x <- as.numeric(unlist(strsplit(line, ' ')))
    total <- total + x[-4] * x[4]
}
if (all(norm - total < 1e-7)) cat('YES') else cat('NO')
```
