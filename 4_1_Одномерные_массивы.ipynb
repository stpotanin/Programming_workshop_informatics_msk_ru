````r
# @title https://stepik.org/lesson/427799/step/15 вариант perplexity.ai
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
# lines <- c('5 1 3 3 3 2 2') # 3
# lines <- c('9 7 3 5 5 5 3 3 8 9') # 6
# lines <- c('8 3 3 3 1 1 1 3 3') # 6
lines <- c('8 3 3 1 1 1 3 3 3') # 8
# lines <- c('1 1 2 7 7 6 5 5 5 6 6 7 7') # 10
# lines <- c('3 3 3 3') # 4
# lines <- c('3 3') # 0

x <- unlist(strsplit(lines[1], ' '))
n <- x[1]
balls <- x[-1]

total_destroyed <- 0
repeat {
  n <- length(balls)
  to_delete <- rep(FALSE, n)

  i <- 1
  found <- FALSE
  while(i <= n) {
    j <- i + 1
    while(j <= n && balls[j] == balls[i]) {
      j <- j + 1
    }
    length_seq <- j - i
    if(length_seq >= 3 && !found) {
      to_delete[i:(j-1)] <- TRUE
      found <- TRUE
      break
    }
    i <- j
  }

  if(!found) break

  destroyed_count <- sum(to_delete)
  total_destroyed <- total_destroyed + destroyed_count
  balls <- balls[!to_delete]
}

cat(total_destroyed)
````
