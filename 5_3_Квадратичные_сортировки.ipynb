```r
# @title https://stepik.org/lesson/263806/step/2 Сортировка выбором максимума
lines <- readLines(file("stdin"))

# для проверки
# lines <- c('2', '3 1') # 1 3
# lines <- c('7', '333 4 55 101 3 44 107') # 3 4 44 55 101 107 333

a <- as.integer(unlist(strsplit(lines[2], ' ')))

max_to_the_end <- function(a) {
    i_max <- which(a == max(a))[1]
    v_max <- a[i_max]
    a[i_max] <- a[length(a)]
    a[length(a)] <- v_max
    return(a)
}

for (i in length(a):1) a[1:i] <- max_to_the_end(a[1:i])

cat(a)
```

```r
# @title https://stepik.org/lesson/263806/step/4 Сортировка вставками
lines <- readLines(file("stdin"))

# для проверки
lines <- c('5', '5 4 3 2 1') # 1 2 3 4 5

a <- as.integer(unlist(strsplit(lines[2], ' ')))

insert_sort <- function(a) {
  n <- length(a)
  for (i in 2:n) {
    x <- a[i]
    j <- i - 1
    while (j > 0 && a[j] > x) {
      a[j + 1] <- a[j]
      j <- j - 1
    }
    a[j + 1] <- x
  }
  return(a)
}

cat(insert_sort(a))
```

```r
# @title https://stepik.org/lesson/263806/step/5 Пузырьковая сортировка_0
lines <- readLines(file("stdin"))

# для проверки
# lines <- c('5', '5 4 3 2 1') # 1 2 3 4 5

a <- as.integer(unlist(strsplit(lines[2], ' ')))

bubble_sort <- function(a) {
    for (i in length(a):2) {
        for (j in 1:(i - 1)) {
            if (a[j] > a[j + 1]) {
                tmp <- a[j]
                a[j] <- a[j + 1]
                a[j + 1] <- tmp
            }
        }
    }
    return (a)
}

cat(bubble_sort(a))
```

```r
# @title https://stepik.org/lesson/263806/step/6 Пузырьковая сортировка: количество обменов
lines <- readLines(file("stdin"))

# для проверки
# lines <- c('5', '1 2 3 4 5') # 0

a <- as.integer(unlist(strsplit(lines[2], ' ')))

bubble_sort_count <- function(a) {
    count <- 0
    for (i in length(a):2) {
        for (j in 1:(i - 1)) {
            if (a[j] > a[j + 1]) {
                tmp <- a[j]
                a[j] <- a[j + 1]
                a[j + 1] <- tmp
                count <- count + 1
            }
        }
    }
    return (count)
}

cat(bubble_sort_count(a))
```

```r
# @title https://stepik.org/lesson/263806/step/7 Экспедиция
lines <- readLines(file("stdin"))

# для проверки
# lines <- c('3 4', '5 10', '7 5', '5 5', '6', '9 5 3 5 12 10') # 4
# lines <- c('3 4', '5 10', '7 5', '5 5', '1', '100') # 0

rafts <- as.integer(unlist(strsplit(lines[-c(1, length(lines) - 1, length(lines))], ' ')))
r <- sort(rafts)
passengers <- as.integer(unlist(strsplit(lines[length(lines)], ' ')))
p <- sort(passengers)

i <- 1
j <- 1
while (i <= length(p) && j <= length(r)) {
    if (p[i] <= r[j]) {
        i <- i + 1  # если поместился
    }
      j <- j + 1    # в любом случае
}

cat(i - 1)
```

```r
# @title https://stepik.org/lesson/263806/step/8 Библиотечный метод
# https://informatics.mccme.ru/mod/statements/view3.php?chapterid=1436
# https://it-wiki.com.ru/algorithms/sorting/insertion-sort
lines <- readLines(file("stdin"))

# для проверки
lines <- c('4', '2 1 5 3') # в  столбик 1 2 5 3     1 2 3 5
lines <- c('6', '6 9 2 10 4 3') # в  столбик 1 2 5 3     1 2 3 5
# lines <- c('4', '1 2 3 5') #

a <- as.integer(unlist(strsplit(lines[2], ' ')))

ins_sort_and_print <- function(arr) {
    for (i in 2:length(arr)) {
        arr_copy <- arr * 1
        x <- arr[i]
        j <- i - 1
        while (j >= 1 && x < arr[j]) {
            arr[j + 1] <- arr[j]
            j <- j - 1
        }
        arr[j + 1] <- x
        if (!all(arr_copy == arr)) {
            cat(arr)
            cat('\n')
        }
    }
}

ins_sort_and_print(a)
```

```r
# @title https://stepik.org/lesson/263806/step/8 Библиотечный метод (Пузырёк для системы проверки) TODO
# https://it-wiki.com.ru/algorithms/sorting/bubble-sort
lines <- readLines(file("stdin"))

# для проверки
lines <- c('4', '2 1 5 3') # в  столбик: 1 2 5 3; 1 2 3 5
lines <- c('6', '6 9 2 10 4 3') # есть в комментариях
lines <- c('4', '1 2 3 5')      # не выводит ничего

a <- as.integer(unlist(strsplit(lines[2], ' ')))

bubble_sort_and_print <- function(arr) {
    n <- length(arr)
    for (i in n:2) { # внешний цикл сверху вниз
        for (j in 1:(i - 1)) {
            if (arr[j] > arr[j + 1]) {
                tmp <- arr[j]
                arr[j] <- arr[j + 1]
                arr[j + 1] <- tmp
                cat(arr)
                cat('\n')
            }
        }
    }
}

bubble_sort_and_print(a)
```
