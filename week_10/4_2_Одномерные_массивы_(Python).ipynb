```r
# @title https://stepik.org/lesson/427800/step/11
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
# lines <- c('165 163 160 160 157 157 155 154', '192') # 1
# lines <- c('165 163 160 160 157 157 155 154', '160') # 5

x <- as.integer(unlist(strsplit(lines[1], ' ')))
p <- as.integer(lines[2])
p_i <- 0
for (i in 1:length(x)) {
    if (p <= x[i]) p_i <- i
}

cat(p_i+1)
```

```r
# @title https://stepik.org/lesson/427800/step/11 бинпоиск
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
# lines <- c('165 163 160 160 157 157 155 154', '192') # 1
# lines <- c('165 163 160 160 157 157 155 154', '160') # 5

rbs <- function(a, key){  # правый бинпоиск
    right <- length(a) + 1
    left <- 0
    while (right - left > 1) {
        middle <- (left + right) %/% 2
        if (a[middle] < key) {
            right <- middle
        } else {
            left <- middle
        }
    }
    return (left)
}

x <- as.integer(unlist(strsplit(lines[1], ' ')))
p <- as.integer(lines[2])

cat(rbs(x, p) + 1)
```
