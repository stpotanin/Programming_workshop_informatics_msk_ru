```r
# @title https://stepik.org/lesson/244031/step/1
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
# lines <- c('700', '750') # 2
# lines <- c('700', '700') # 1

n <- as.integer(lines[1])
m <- as.integer(lines[2])

cat((m - 0.5) %/% n + 1)
```

```r
# @title https://stepik.org/lesson/244031/step/2
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
# lines <- c('7', '30') # 5
# lines <- c('7', '7') # 0

n <- as.integer(lines[1])
k <- as.integer(lines[2])

cat((n - k %% n) %% n)
```

```r
# @title https://stepik.org/lesson/244031/step/3
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
# lines <- c('10', '3', '2') # 8
# lines <- c('10', '5', '3') # 4
# lines <- c('10', '10', '5') # 1
# lines <- c('5', '100', '50') # 1
lines <- c('50', '10', '4') # 1

h <- as.integer(lines[1])
a <- as.integer(lines[2])
b <- as.integer(lines[3])

cat((h - b - 1) %% h %/% (a - b) + 1)
```

```r
# @title https://stepik.org/lesson/244031/step/3 циклом
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
# lines <- c('10', '3', '2') # 8
lines <- c('10', '5', '3') # 4
# lines <- c('10', '10', '5') # 1
# lines <- c('10', '20', '19') # 1

h <- as.integer(lines[1])
a <- as.integer(lines[2])
b <- as.integer(lines[3])

n <- 0
d <- 0
while (TRUE) {
    n <- n+1
    d <- d + a
    if (d >= h) {
        break
    } else {
        d <- d - b
    }
}

cat(n)
```

```r
# @title https://stepik.org/lesson/244031/step/4
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
# lines <- c('110') # 1
# lines <- c('-232') # 0

x <- abs(as.integer(lines[1]))
a <- sprintf("%04d", x)
a_rev <- paste0(substr(rep(a, n <- nchar(a)), n:1, n:1), collapse="")

cat((a == a_rev) * 1)
```

```r
# @title https://stepik.org/lesson/244031/step/5
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
# lines <- c('2', '8') # 1
# lines <- c('2', '7') # 0

n <- max(as.integer(lines[1]), as.integer(lines[2]))
m <- min(as.integer(lines[1]), as.integer(lines[2]))

cat((n / m == n %/% m) * 1)
```

```r
# @title https://stepik.org/lesson/244031/step/6
# https://informatics.msk.ru/mod/statements/view.php?chapterid=2958
lines <- readLines(file('stdin'))

# для проверки
# lines <- c('7', '14') # 14
# lines <- c('5', '5') # 5
# lines <- c('3', '1') # 3

a <-as.integer(lines[1])
b <- as.integer(lines[2])

is_equal = a %% b == 0 && b %% a == 0

cat((a + b) - (a %% b * b %/% a + b %% a * a %/% b) %/% (a %/% b + b %/% a) - a * is_equal)



# cat((a + b) - (a %% b * b %/% a + b %% a * a %/% b) %/% (a %/% b + b %/% a))
```

```r
# Вариант (классическое решение)
lines <- readLines(file('stdin'))

# для проверки
lines <- c('7', '14') # 14
lines <- c('5', '5') # 5
lines <- c('3', '1') # 3

a <-as.integer(lines[1])
b <- as.integer(lines[2])
cat((((a %/% b) * a) + ((b %/% a) * b)) %/% ((a %/% b) + (b %/% a)))
```
