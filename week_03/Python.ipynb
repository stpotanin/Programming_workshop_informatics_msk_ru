```python
# lines <- c('01.0001.100.001') # 0

def is_correct(ip: str)-> bool:
    ip = ip.split('.')
    if len(ip) != 4:
        return False
    for a in ip:
        if len(a) > 3:
            return False
        if not a.isdigit():
            return False
        if not 0 <= int(a) < 256:
            return False
    return True

print(int(is_correct(input())))
```

```python
# @title https://stepik.org/lesson/427795/step/1
from itertools import groupby

s, n = input().split()

# для проверки
# s, n = '1 4'.split() # ответ 1211

s = [x for x in s]
n = int(n)

# Функция для группировки:
gr = lambda x: [list(group) for key, group in groupby(x)]

for i in range(0, n-1):
    s = gr(s)
    new_s =[]
    for ss in s:
        number, lenght = ss[0], len(ss)
        new_s.append(str(lenght))
        new_s.append(number)
    s = new_s

print(*s, sep='')
```

```python
# Генерация n случайных точек в диапазоне от min_val до max_val
import random

def generate_points(n, min_val=-1000, max_val=1000):
    points = [(random.uniform(min_val, max_val), random.uniform(min_val, max_val)) for _ in range(n)]
    return points

# Пример:
points = generate_points(10000)  # 10 000 случайных точек

#############

import sys
import math

# для локальной проверки:
# lines = ['4', '1 2', '2 3', '44 55', '66 77'] # 99.2471662063
lines = ['2', '1 2', '2 3']                 # 1.4142135623

# для stdin:
# lines = sys.stdin.read().strip().split('\n')

n = int(lines[0])
points = [list(map(float, line.split())) for line in lines[1:]]

distances = [math.sqrt((points[i][0] - points[j][0])**2 + (points[i][1] - points[j][1])**2)
             for i in range(n) for j in range(i+1, n)]

max_dist = max(distances)
max_formated = int(max_dist * 10**12) / 10**12 # обрезаем
print(max_formated)
```

```python
class Student:
    def __init__(self, data: list[str]):
        self.name = ' '.join(data[:2])
        self.mark = sum(map(int, data[2:]))


best = [0, []]
for _ in range(int(input())):
    stud = Student(input().split())
    if stud.mark > best[0]:
        best = [stud.mark, []] # ПЕРЕписываем весь список
        print(best)
    if stud.mark == best[0]:
        best[1].append(stud.name)  # ДОписываем
        print(best)

print('\n'.join(best[1]))

# 4
# Belova Galina 4 3 5
# Babat Anna 5 4 5
# Moroz Yaroslav 4 5 5
# Lev Barhudarov 5 5 5
```
