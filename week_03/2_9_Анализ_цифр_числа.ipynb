```r
# @title https://stepik.org/lesson/244350/step/4
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
lines <- c('6') # 011

# Функция для получения бинарного представления
to_binary <- function(n) {
    if (n == 0) return(0)
    bits <- c()
    while (n > 0) {
        bits <- c(n %% 2, bits)
    n <- n %/% 2
    }
    return(bits)
}

cat(rev(to_binary(as.integer(lines)[1])), sep='')
```

```r
# @title https://stepik.org/lesson/244350/step/6
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
lines <- c('100') # 18

k <- as.integer(lines[1])

palindrome <- function(x) as.integer(paste(rev(unlist(strsplit(as.character(x), ''))), collapse=''))

n <- 0
for (i in 1:k) {
    if (palindrome(i) == i) {
        n <- n + 1
    }
}

paste(n)
```

```r
# @title https://stepik.org/lesson/244350/step/6 вариант
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
# lines <- c('100') # 18

k <- as.integer(lines[1])

palindrome <- function(x) as.integer(paste(rev(unlist(strsplit(as.character(x), ''))), collapse=''))

cat(sum(sapply(1:k, function(x) palindrome(x) == x)))
```
