```r
# @title https://stepik.org/lesson/427794/step/9
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
# lines <- c('caba', 'ababacaba') # yes

# делаем слова списками
w1 <- unlist(strsplit(lines[1], ''))
w2 <- unlist(strsplit(lines[2], ''))

n1 <- length(w1)    # длина 1-го слова
n2 <- length(w2)    # длина 2-го слова

res <- 'no'

# накладываем первое слово на второе со сдвигом
for (i in 1:(n2-n1+1)) if (all(w1 == w2[i:(i+n1-1)])) res <- 'yes'

cat(res)
```

```r
# @title https://stepik.org/lesson/427794/step/10
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
# lines <- c('South 19', 'East 5') # 5 -19

a <- list(North = c(0, 1), South = c(0, -1), East = c(1, 0), West = c(-1, 0))

x <- 0
y <- 0

for (line in lines) {
    step <- unlist(strsplit(line, ' '))
    direction <- step[1]
    distance <- as.integer(step[2])
    x <- x + a[[direction]][1] * distance
    y <- y + a[[direction]][2] * distance
}

cat(x, y)
```

```r
# @title https://stepik.org/lesson/427794/step/12
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
lines <- c('XPSE', '1') # WORD
# lines <- c('ZABC', '3') # WXYZ

cipher <- unlist(strsplit(lines[1], split = ''))
gag <- as.integer(lines[2])
word <- c()
n <- length(LETTERS)
for (c in cipher) {
    w <- LETTERS[((match(c, LETTERS) - gag - 1) %% n) + 1]
    word <- c(word, w)
}
cat(word, sep='')
```

```r
# @title https://stepik.org/lesson/427794/step/15
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
# lines <- c('127.0.0.1') # 1
# lines <- c('127.0.x.1') # 0
# lines <- c('10.01.100.001') # 1
lines <- c('01.000.100.001') # 0

s <- unlist(strsplit(lines[1], split = '\\.'))

res <- 1

if (length(s) != 4) {
  res <- 0
} else {
  for (ss in s) {
    # проверка только цифры
    if (!grepl("^[0-9]+$", ss)) {
      res <- 0
      break
    }
    # три и более ведущих нулей
    if (nchar(ss) > 1 && substr(ss, 1, 3) == "000") {
      res <- 0
      break
    }
    # диапазон от 0 до 255
    num <- as.integer(ss)
    if (is.na(num) || num < 0 || num > 255) {
      res <- 0
      break
    }
  }
}

cat(res)
```

```r
# @title https://stepik.org/lesson/427794/step/16
input <- file("stdin")
lines <- readLines(input)
close(input)

# для проверки
# lines <- c('ABAC') # ABA

s <- unlist(strsplit(lines[1], split = ''))
res <- s[1]

for (i in 1:(length(s))) {
    for(j in (i:length(s))) {
        if (all(s[i:j] == s[j:i]) & length(s[i:j]) > length(res)) {
            res <- s[i:j]
        }
    }
}

cat(res, sep='')
```
