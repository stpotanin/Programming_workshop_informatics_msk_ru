```r
# @title https://stepik.org/lesson/263809/step/1 Сортировка подсчетом
# https://ru.wikipedia.org/wiki/%D0%A1%D0%BE%D1%80%D1%82%D0%B8%D1%80%D0%BE%D0%B2%D0%BA%D0%B0_%D0%BF%D0%BE%D0%B4%D1%81%D1%87%D1%91%D1%82%D0%BE%D0%BC

simple_counting_sort <- function(arr) {
    # нормируем список в диапазон, начинающийся с 1
    bias  <- min(arr) - 1
    arr  <- arr - bias
    count <- rep(0, max(arr)) # список из нулей длиной в максимальный элемент
    for (i in 1:length(arr)) {
        # когда встречаем число в списке,
        # увеличиваем значение по соответствующему адресу на 1
        count[arr[i]] = count[arr[i]] + 1
    }
    res <- c()
    for (j in 1:max(arr)) {
        # каждый индекс добавляем в спиок соответствующее число раз
        res <- c(res, rep(j, count[j]))
    }
    # результату возвращем смещение, полученное при нормировании
    return (res + bias)
}

lines <- readLines(file("stdin"))

# для проверки
# lines <- c('5', '1 3 4 3 5') # 1 3 3 4 5
# lines <- c('5', '1 3 -4 3 5') # -4 1 3 3 5

arr <- as.integer(unlist(strsplit(lines[2], ' ')))

cat(simple_counting_sort(arr))
```

```r
# @title https://stepik.org/lesson/263809/step/2 Носки

lines <- readLines(file("stdin"))

# для проверки
# lines <- c('39 4 7',
#            '3 21',
#            '3 15',
#            '2 20',
#            '3 17',
#            '4', '17', '33', '5', '9', '25', '37'
#            ) # 4 3 0 4 4 0 0 в столбик

lnm <- as.integer(unlist(strsplit(lines[1], ' ')))
l <- lnm[1]
n <- lnm[2]
m <- lnm[3]

socks <- matrix(as.integer(unlist(strsplit(lines[2:(n + 1)], ' '))), ncol = 2, byrow = TRUE)
points <- as.integer(unlist(strsplit(lines[-c(1:(n + 1))], ' ')))

board <- rep(0, l)
for (i in 1:n) {
    # плюс 1 по интервалу, занятому носком
    board[socks[i, 1]:socks[i, 2]] <- board[socks[i, 1]:socks[i, 2]] + 1
}

# вывод результата в каждой из точек
for (p in points) cat(board[p], sep='', end='\n')
```

```r
# @title https://stepik.org/lesson/263809/step/3 Анаграммы

lines <- readLines(file("stdin"))

# для проверки
# lines <- c('sharm', 'marsh') # YES

w1 <- unlist(strsplit(lines[1], ''))
w2 <- unlist(strsplit(lines[2], ''))

abc <- rep(0, 26)

# считаем буквы первого слова
for (w in w1) {
    abc[which(letters == w)] <- abc[which(letters == w)] + 1
}

# вычитаем буквы второго слова
for (w in w2) {
    abc[which(letters == w)] <- abc[which(letters == w)] - 1
}

cat(ifelse(all(abc == 0), 'YES', 'NO'))
```

```r
# @title https://stepik.org/lesson/263809/step/5 Клавиатура

lines <- readLines(file("stdin"))

# для проверки
# lines <- c('5', '1 50 3 4 3', '16', '1 2 3 4 5 1 3 3 4 5 5 5 5 5 4 5') # yes no no no yes в столбик

resurs <- as.integer(unlist(strsplit(lines[2], ' ')))
keystrokes <- as.integer(unlist(strsplit(lines[4], ' ')))

for (k in keystrokes) {
    resurs[k] <- resurs[k] - 1
}

for (r in resurs) {
    cat(ifelse(r < 0, 'yes', 'no'))
    cat('\n')
}
```

```r
# @title https://stepik.org/lesson/263809/step/7 Палиндром_0
lines <- readLines(file("stdin"))

# для проверки
lines <- c('3', 'AABB') # ABA

word <- unlist(strsplit(lines[2], ''))

# подсчёт букв в слове
counter <- rep(0, 26)
for (x in word) {
    i <- grep(x, LETTERS) # индекс буквы в алфавите
    counter[i] <- counter[i] + 1
}

# лепим левую и правую половины палиндрома в алфавитном порядке
left <- c()
right <- c()
for (i in 1:26) {
    n <- counter[i] %/% 2 # число пар текущей буквы
    left <- c(left, rep(LETTERS[i], n))
    right <- c(rep(LETTERS[i], n), right)
}

# если надо, вставляем серединку
medium <- LETTERS[counter %% 2 == 1][1]
if (is.na(medium)) {
  palindrome <- c(left, right)
} else {
  palindrome <- c(left, medium, right)
}

cat(palindrome, sep='')
medium
```
