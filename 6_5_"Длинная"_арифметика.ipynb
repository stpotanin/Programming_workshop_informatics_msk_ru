```r
# @title https://stepik.org/lesson/464320/step/7 Задача №136. Умножение на длинное
# https://informatics.msk.ru/mod/statements/view.php?chapterid=136
ptm <- proc.time() # TODO
# Сложение, функция принимает списки из цифр
str_add <- function(x, y) {
    if (length(x) < length(y)) {
        tmp <- x
        x <- y
        y <- tmp
    }

    # дописываем нули слева к короткому числу
    y <-  c(rep(0, length(x) - length(y)), y)

    # вычисляем значения суммы поразрядно
    carry <- 0
    for (i in length(x):1) {
        s <- (x[i] + y[i] + carry) %% 10
        carry <- (x[i] + y[i] + carry) %/% 10
        x[i] <- s
    }

    # дополнительный разряд, если необходимо
    if (carry != 0) x <- c(c(1), x)
    x   # результат
}

# Вычитание со знаком, функция принимает списки из цифр
str_minus <- function(x, y) {
    minus <- FALSE
    if (length(x) < length(y)) {
        tmp <- x
        x <- y
        y <- tmp
        minus <- TRUE
    }
    # дописываем нули слева к вычитаемому
    y <-  c(rep(0, length(x) - length(y)), y)
    if (all(x == y)) return(0)
    # справа налево
    borrow <- 0
    for (i in length(x):1) {
        d <- (x[i] - y[i] + borrow) %% 10
        borrow <- (x[i] - y[i] + borrow) %/% 10
        x[i] <- d
    }
    if (borrow) {
        x <- longMinus(c(1, rep(0, length(x))), x)
        minus <- TRUE
    }

    # убираем ведущие нули
    i <- which(x != 0)[1]
    x <- x[i:length(x)]
    if (minus) x <- c('-', x)
    return(x)
}

# Функция для дополнения нулями слева
zfill <- function(x, width) {
  formatC(x, width = width, flag = "0")
}

# строковое представление a и b, BASE=10. Обходит ограничение R на 31 цифру.
karatsuba_decimal <- function(str_a, str_b) {
    # Базовый случай (одно из чисел меньше 10 цифр)
    if (length(str_a) < 5 | length(str_b) < 5) {
        if (length(str_a) < length(str_b)) {
            tmp <- str_b
            str_b <- str_a
            str_a <- tmp
        }
        b <- as.integer(paste(str_b, collapse = ''))
        res <- c(0)
        for (i in 1:b) {
            res <- str_add(res, str_a)
        }
        return (res)
    }

    m <- max(length(str_a), length(str_b))
    m2 <- (m + 1) %/% 2

    # Разделение строк (строчное быстрее чем деление)
    # high: первые N символов
    high1 <- head(str_a, max(0, length(str_a)-m2))
    high2 <- head(str_b, max(0, length(str_b)-m2))

    # low: последние M символов
    low1 <- tail(str_a, length(str_a)-m2+1)
    low2 <- tail(str_b, length(str_b)-m2+1)

    # Рекурсивные вызовы
    z0 = karatsuba_decimal(low1, low2)
    z1 = karatsuba_decimal(str_add(low1, high1), str_add(low2, high2))
    z2 = karatsuba_decimal(high1, high2)

    # Формируем результат, формула Карацубы: z2*10^(m2) + (z1-z2-z0)*10^m + z0
    p1 <- c(z2, rep(0L, 2 * m2))  # z2 * 10^(2*m2)
    mid <- str_minus(str_minus(z1, z2), z0)
    p2 <- c(mid, rep(0L, m2))     # mid * 10^m2, где mid = z1 - z2 - z0
    p3 <- z0

    result <- as.character(as.integer(p1) + as.integer(p2 + m2 * '0') + as.integer(p3))
    return (result)
}

lines <- readLines(file('stdin'))

# проверка
lines <- c('999999', '999999') # 998001 (11 секунд)
# lines <- c('9999999', '9999999') # 99999980000001 (125 секунд)
# lines <- c('9999999999999999999999999999999999999999', '9999999999999999999999999999999999999999') # 0

x <- as.integer(strsplit(lines[1], '')[[1]])
y <- as.integer(strsplit(lines[2], '')[[1]])
cat(karatsuba_decimal(x, y), sep = '')

proc.time() - ptm

# TODO
```

```r
# @title https://stepik.org/lesson/464320/step/14 Задача №143. Факториал
# https://informatics.msk.ru/mod/statements/view.php?chapterid=143

str_add <- function(x, y) {
    if (length(x) < length(y)) {
        tmp <- x
        x <- y
        y <- tmp
    }

    # дописываем нули слева к короткому числу
    y <-  c(rep(0, length(x) - length(y)), y)

    # вычисляем значения суммы поразрядно
    carry <- 0
    for (i in length(x):1) {
        s <- (x[i] + y[i] + carry) %% 10
        carry <- (x[i] + y[i] + carry) %/% 10
        x[i] <- s
    }

    # дополнительный разряд, если необходимо
    if (carry != 0) x <- c(c(1), x)
    x   # результат
}

# факториал сложением строк
str_factorial <- function(x) {
  if (x <= 1) return(1)
  prev <- str_factorial(x - 1)
  res <- c(0)
  for (i in 1:x) {
    res <- str_add(res, prev)
  }
  return(res)
}


lines <- readLines(file('stdin'))

# проверка
# lines <- c('1') # 1
# lines <- c('5') # 120
lines <- c('30') # 479001600

x <- as.integer(lines[1])
cat(str_factorial(x), sep = '')
```

```r
# @title https://stepik.org/lesson/464320/step/15 Задача №144. N^N mod 10^P
# https://informatics.msk.ru/mod/statements/view.php?chapterid=145

rbs(key):  # правый бинпоиск
    right = 10*500
    left = 0
    while right - left > 1:
        middle = (left + right) // 2
        # print(middle)
        if middle * middle > key:
            right = middle
            # print("Сдвигаем right")
        else:
            left = middle
            # print("Сдвигаем left")
    return left


a = 151257600
print(rbs(a))
print('Точно:', a**.5)
```
